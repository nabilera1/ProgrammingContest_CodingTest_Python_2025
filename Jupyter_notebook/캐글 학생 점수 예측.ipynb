```python
# 1. 필요한 라이브러리 설치 (터미널 또는 노트북 첫 셀에서)
# pip install kagglehub pandas matplotlib

import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# 2. KaggleHub를 이용해 데이터셋 다운로드
# "username/dataset-name" 형식
path = kagglehub.dataset_download("spscientist/students-performance-in-exams")
print("데이터셋이 저장된 폴더 경로:", path)

# 3. 실제 CSV 파일 불러오기
# 폴더 안에 어떤 파일이 있는지는 직접 확인해도 되고,
# 보통 이 데이터셋은 "StudentsPerformance.csv"라는 파일을 사용함.
csv_path = path + "/StudentsPerformance.csv"
df = pd.read_csv(csv_path)

# 4. 데이터 구조 확인
print("\n[행, 열 개수]:", df.shape)
print("\n[앞부분 5행]")
print(df.head())
```
