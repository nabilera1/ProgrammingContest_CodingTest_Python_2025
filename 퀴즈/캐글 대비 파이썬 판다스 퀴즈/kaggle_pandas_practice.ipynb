### 문제 8: 이상치 제거 (IQR 활용)
'Value' 컬럼에 대해 IQR 방법을 사용하여 이상치를 제거하시오. (Q1-1.5*IQR ~ Q3+1.5*IQR)

> **힌트:** quantile()로 Q1, Q3를 구하고 범위를 설정하여 필터링하세요.

<details style="background-color: #ffffff; border: 1px solid #e0e0e0; padding: 10px; border-radius: 5px;">
<summary style="font-weight: bold; color: #1e3a8a;">👉 정답 및 해설 보기 (클릭)</summary>

```python
Q1 = df['Value'].quantile(0.25)
Q3 = df['Value'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_filtered = df[(df['Value'] >= lower_bound) & (df['Value'] <= upper_bound)]
print(df_filtered)
```

**💡 설명:**
IQR 방식은 데이터의 분포를 기반으로 통계적으로 이상치를 탐지하는 일반적인 방법입니다.
