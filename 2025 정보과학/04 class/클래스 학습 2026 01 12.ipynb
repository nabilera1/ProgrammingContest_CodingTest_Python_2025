## 1ë‹¨ê³„: í´ë˜ìŠ¤ì™€ ê°ì²´ì˜ ê¸°ë³¸ (Basic Structure)

í•µì‹¬ ê°œë…: \_\_init__ (ìƒì„±ì), self (ì¸ìŠ¤í„´ìŠ¤ ìì‹ ), ì¸ìŠ¤í„´ìŠ¤í™”

```python
class User:
    # __init__: ê°ì²´ê°€ ìƒì„±ë  ë•Œ ìë™ìœ¼ë¡œ ì‹¤í–‰ë˜ì–´ ì†ì„±ì„ ì´ˆê¸°í™”í•¨
    def __init__(self, name, age):
        self.name = name  # ì†ì„±(Attribute) ì •ì˜
        self.age = age

    # ë©”ì„œë“œ(Method): í´ë˜ìŠ¤ ë‚´ë¶€ì— ì •ì˜ëœ í•¨ìˆ˜
    def introduce(self):
        return f"ì•ˆë…•í•˜ì„¸ìš”, ì €ëŠ” {self.name}ì´ê³  {self.age}ì‚´ì…ë‹ˆë‹¤."

# --- ì‹¤í–‰ ---
# ê°ì²´(ì¸ìŠ¤í„´ìŠ¤) ìƒì„±
user1 = User("ê¹€ì² ìˆ˜", 25)
user2 = User("ì´ì˜í¬", 30)

print(user1.introduce())  # ì¶œë ¥: ì•ˆë…•í•˜ì„¸ìš”, ì €ëŠ” ê¹€ì² ìˆ˜ì´ê³  25ì‚´ì…ë‹ˆë‹¤.
print(user2.name)         # ì¶œë ¥: ì´ì˜í¬
```

## 2ë‹¨ê³„: ìº¡ìŠí™”ì™€ ì ‘ê·¼ ì œì–´ (Encapsulation)
ë°ì´í„°ë¥¼ ë³´í˜¸í•˜ê³  ë¬´ê²°ì„±ì„ ìœ ì§€í•˜ê¸° ìœ„í•´ ë³€ìˆ˜ë¥¼ ì§ì ‘ ê±´ë“œë¦¬ì§€ ëª»í•˜ê²Œ í•˜ëŠ” ë‹¨ê³„ì…ë‹ˆë‹¤.

í•µì‹¬ ê°œë…: _ë³€ìˆ˜ëª… (ë‚´ë¶€ìš© ë³€ìˆ˜ ì•”ì‹œ), @property (ê²Œí„°), @ë³€ìˆ˜ëª….setter (ì„¸í„°)

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  # __ë³€ìˆ˜ëª…: ë¹„ê³µê°œ(Private) ë³€ìˆ˜ë¡œ ë§Œë“¦

    # @property: ë©”ì„œë“œë¥¼ ë³€ìˆ˜ì²˜ëŸ¼ ì ‘ê·¼í•˜ê²Œ í•´ì¤Œ (Getter)
    @property
    def balance(self):
        return self.__balance

    # ë°ì´í„°ë¥¼ ë³€ê²½í•  ë•Œ ê²€ì¦ ë¡œì§ì„ ì¶”ê°€ (Setter)
    @balance.setter
    def balance(self, amount):
        if amount < 0:
            print("ì”ì•¡ì€ ìŒìˆ˜ê°€ ë  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        else:
            self.__balance = amount

    def deposit(self, amount):
        self.__balance += amount
        print(f"{amount}ì› ì…ê¸ˆ ì™„ë£Œ. í˜„ì¬ ì”ì•¡: {self.__balance}ì›")

# --- ì‹¤í–‰ ---
my_account = BankAccount("í™ê¸¸ë™", 1000)

# my_account.__balance = 5000  # ì—ëŸ¬ ë°œìƒ! (ì™¸ë¶€ì—ì„œ ì§ì ‘ ì ‘ê·¼ ë¶ˆê°€)
my_account.deposit(500)        # ë©”ì„œë“œë¥¼ í†µí•´ ì•ˆì „í•˜ê²Œ ìˆ˜ì •
print(my_account.balance)      # @propertyë¥¼ í†µí•´ ë³€ìˆ˜ì²˜ëŸ¼ ì½ê¸° ê°€ëŠ¥
```

## ì¶”ê°€ : 2ë‹¨ê³„: ìº¡ìŠí™”ì™€ ì ‘ê·¼ ì œì–´ (ì¶”ê°€ ì˜ˆì œ: ì„±ì  ê´€ë¦¬)
Getterì™€ SetterëŠ” "ë°ì´í„°ê°€ ì´ìƒí•˜ê²Œ ì…ë ¥ë˜ëŠ” ê²ƒì„ ë§‰ì„ ë•Œ" ê°€ì¥ ë§ì´ ì”ë‹ˆë‹¤. ì´ë²ˆì—ëŠ” 0~100ì  ì‚¬ì´ì˜ ì ìˆ˜ë§Œ ì…ë ¥ë°›ë„ë¡ í†µì œí•˜ëŠ” Student í´ë˜ìŠ¤ë¥¼ ì˜ˆë¡œ ë“¤ê² ìŠµë‹ˆë‹¤.

```python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.__score = score  # ë¹„ê³µê°œ ë³€ìˆ˜ (ì§ì ‘ ì ‘ê·¼ ë¶ˆê°€)

    # Getter: ì ìˆ˜ë¥¼ ì½ì–´ì˜¬ ë•Œ í˜¸ì¶œ
    @property
    def score(self):
        return self.__score

    # Setter: ì ìˆ˜ë¥¼ ìˆ˜ì •í•  ë•Œ í˜¸ì¶œ (ìœ íš¨ì„± ê²€ì‚¬ ë¡œì§ í¬í•¨)
    @score.setter
    def score(self, new_score):
        if 0 <= new_score <= 100:
            self.__score = new_score
            print(f"{self.name} í•™ìƒì˜ ì ìˆ˜ê°€ {new_score}ì ìœ¼ë¡œ ìˆ˜ì •ë˜ì—ˆìŠµë‹ˆë‹¤.")
        else:
            print(f"[ì˜¤ë¥˜] {new_score}ì ì€ ìœ íš¨í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤. (0~100ì  ì‚¬ì´ë§Œ ê°€ëŠ¥)")

# --- ì‹¤í–‰ ---
s1 = Student("ë°•ê³µë¶€", 80)

# 1. Getter ì‚¬ìš© (ë³€ìˆ˜ì²˜ëŸ¼ ì½ê¸°)
print(f"í˜„ì¬ ì ìˆ˜: {s1.score}")

# 2. Setter ì‚¬ìš© (ì •ìƒ ê°’ ì…ë ¥)
s1.score = 95  # ë‚´ë¶€ì ìœ¼ë¡œ @score.setterê°€ ì‹¤í–‰ë¨ -> ì„±ê³µ

# 3. Setter ì‚¬ìš© (ì˜ëª»ëœ ê°’ ì…ë ¥ - ë°ì´í„° ë³´í˜¸)
s1.score = 150 # ë‚´ë¶€ì ìœ¼ë¡œ ê²€ì¦ ë¡œì§(ifë¬¸)ì— ê±¸ë¦¼ -> ì‹¤íŒ¨ ë©”ì‹œì§€ ì¶œë ¥
s1.score = -10 # ì‹¤íŒ¨

print(f"ìµœì¢… ì ìˆ˜: {s1.score}") # ì—¬ì „íˆ 95ì  ìœ ì§€
```

## 3ë‹¨ê³„: ìƒì†ê³¼ ì˜¤ë²„ë¼ì´ë”© (Inheritance)
ê¸°ì¡´ ì½”ë“œë¥¼ ì¬ì‚¬ìš©í•˜ê³  í™•ì¥í•˜ëŠ” ê°ì²´ì§€í–¥ì˜ ê½ƒì…ë‹ˆë‹¤.

í•µì‹¬ ê°œë…: class ìì‹(ë¶€ëª¨), super(), ë©”ì„œë“œ ì˜¤ë²„ë¼ì´ë”©(ë®ì–´ì“°ê¸°)

```python
# ë¶€ëª¨ í´ë˜ìŠ¤
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("í•˜ìœ„ í´ë˜ìŠ¤ì—ì„œ êµ¬í˜„í•´ì•¼ í•©ë‹ˆë‹¤.")

# ìì‹ í´ë˜ìŠ¤ 1
class Dog(Animal):
    def speak(self):  # ë¶€ëª¨ì˜ speakë¥¼ ì¬ì •ì˜ (Overriding)
        return "ë©ë©!"

# ìì‹ í´ë˜ìŠ¤ 2
class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name)  # ë¶€ëª¨ì˜ __init__ì„ í˜¸ì¶œí•˜ì—¬ name ì´ˆê¸°í™”
        self.color = color      # Catë§Œì˜ ê³ ìœ  ì†ì„± ì¶”ê°€

    def speak(self):
        return "ì•¼ì˜¹~"

# --- ì‹¤í–‰ ---
dog = Dog("ë°”ë‘‘ì´")
cat = Cat("ë‚˜ë¹„", "í°ìƒ‰")

print(f"{dog.name}: {dog.speak()}")  # ë°”ë‘‘ì´: ë©ë©!
print(f"{cat.name} ({cat.color}): {cat.speak()}") # ë‚˜ë¹„ (í°ìƒ‰): ì•¼ì˜¹~
```

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"[ì…ê¸ˆ] {amount}ì›. ì”ì•¡: {self.balance}ì›")

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"[ì¶œê¸ˆ] {amount}ì›. ì”ì•¡: {self.balance}ì›")
        else:
            print(f"[ì¶œê¸ˆ ì‹¤íŒ¨] ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤. (í˜„ì¬ ì”ì•¡: {self.balance}ì›)")

# 1. ìƒì† ì˜ˆì œ: ì´ì ê¸°ëŠ¥ì´ ì¶”ê°€ëœ 'ì ê¸ˆ í†µì¥'
class SavingsAccount(BankAccount):
    def __init__(self, owner, balance, interest_rate):
        # super()ë¥¼ ì‚¬ìš©í•´ ë¶€ëª¨(BankAccount)ì˜ __init__ì„ ê·¸ëŒ€ë¡œ ì‚¬ìš©
        super().__init__(owner, balance)
        self.interest_rate = interest_rate  # ì´ììœ¨ ì†ì„± ì¶”ê°€

    # ë¶€ëª¨ì—ëŠ” ì—†ëŠ” ìƒˆë¡œìš´ ë©”ì„œë“œ ì¶”ê°€ (ê¸°ëŠ¥ í™•ì¥)
    def add_interest(self):
        interest = int(self.balance * self.interest_rate)
        self.balance += interest
        print(f"[ì´ì ì§€ê¸‰] {interest}ì›ì´ ì§€ê¸‰ë˜ì—ˆìŠµë‹ˆë‹¤. ì”ì•¡: {self.balance}ì›")

# 2. ì˜¤ë²„ë¼ì´ë”© ì˜ˆì œ: ì”ì•¡ì´ ì—†ì–´ë„ ì¶œê¸ˆë˜ëŠ” 'ë§ˆì´ë„ˆìŠ¤ í†µì¥'
class MinusAccount(BankAccount):
    def __init__(self, owner, balance, limit):
        super().__init__(owner, balance)
        self.limit = limit  # ë§ˆì´ë„ˆìŠ¤ í•œë„

    # ë¶€ëª¨ì˜ withdraw ë©”ì„œë“œë¥¼ ë®ì–´ì“°ê¸° (Method Overriding)
    def withdraw(self, amount):
        # ì”ì•¡ + í•œë„ ë‚´ì—ì„œ ì¶œê¸ˆ ê°€ëŠ¥í•˜ë„ë¡ ë¡œì§ ë³€ê²½
        if self.balance + self.limit >= amount:
            self.balance -= amount
            print(f"[ë§ˆì´ë„ˆìŠ¤ ì¶œê¸ˆ] {amount}ì›. ì”ì•¡: {self.balance}ì›")
        else:
            print(f"[ì¶œê¸ˆ ì‹¤íŒ¨] í•œë„ ì´ˆê³¼ì…ë‹ˆë‹¤. (ê°€ëŠ¥ì•¡: {self.balance + self.limit}ì›)")

# --- ì‹¤í–‰ ---
print("--- ì ê¸ˆ í†µì¥ í…ŒìŠ¤íŠ¸ ---")
acc1 = SavingsAccount("ê¹€ì„±ì‹¤", 10000, 0.05) # ì´ììœ¨ 5%
acc1.add_interest()  # ë¶€ëª¨ì—” ì—†ê³  ìì‹ì—ê²Œë§Œ ìˆëŠ” ê¸°ëŠ¥

print("\n--- ë§ˆì´ë„ˆìŠ¤ í†µì¥ í…ŒìŠ¤íŠ¸ ---")
acc2 = MinusAccount("ì´ììœ ", 0, 5000) # ì”ì•¡ 0ì›, í•œë„ 5000ì›
acc2.withdraw(3000)  # ë¶€ëª¨ë¼ë©´ ì‹¤íŒ¨í–ˆê² ì§€ë§Œ, ì˜¤ë²„ë¼ì´ë”© ë•ë¶„ì— ì„±ê³µ (-3000ì›)
acc2.withdraw(3000)  # í•œë„(-5000ì›)ë¥¼ ë„˜ìœ¼ë¯€ë¡œ ì‹¤íŒ¨
```

## 4ë‹¨ê³„: ë§¤ì§ ë©”ì„œë“œ í™œìš© (Magic Methods)
íŒŒì´ì¬ì˜ ë‚´ì¥ í•¨ìˆ˜ë‚˜ ì—°ì‚°ì(+, -, print, len)ê°€ ë‚´ í´ë˜ìŠ¤ì—ì„œë„ ì‘ë™í•˜ê²Œ ë§Œë“­ë‹ˆë‹¤. ë°ì´í„° ë¶„ì„ì´ë‚˜ ìˆ˜í•™ ê´€ë ¨ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì œì‘ ì‹œ í•„ìˆ˜ì…ë‹ˆë‹¤.

í•µì‹¬ ê°œë…: \_\_str__ (ë¬¸ìì—´ í‘œí˜„), \_\_add__ (ë§ì…ˆ ì—°ì‚°ì), \_\_eq__ (ë¹„êµ ì—°ì‚°ì)

```python
class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # print() í•¨ìˆ˜ í˜¸ì¶œ ì‹œ ë³´ì—¬ì¤„ ë¬¸ìì—´ ì •ì˜
    def __str__(self):
        return f"Vector({self.x}, {self.y})"

    # + ì—°ì‚°ì ì •ì˜ (Vectorë¼ë¦¬ ë”í•˜ê¸°)
    # ìˆ˜í•™ì  ì •ì˜: $(x_1, y_1) + (x_2, y_2) = (x_1+x_2, y_1+y_2)$
    def __add__(self, other):
        return Vector2D(self.x + other.x, self.y + other.y)

    # == ì—°ì‚°ì ì •ì˜
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

# --- ì‹¤í–‰ ---
v1 = Vector2D(3, 4)
v2 = Vector2D(1, 2)
v3 = v1 + v2  # __add__ ìë™ í˜¸ì¶œ

print(v1)        # Vector(3, 4) -> __str__ í˜¸ì¶œ
print(v3)        # Vector(4, 6)
print(v1 == v2)  # False -> __eq__ í˜¸ì¶œ
```

## 5ë‹¨ê³„: ëª¨ë˜ íŒŒì´ì¬ íŒ¨í„´ (Dataclass & Classmethod)
ìµœì‹  íŒŒì´ì¬(3.7+) ì‹¤ë¬´ì—ì„œ ê°€ì¥ ë§ì´ ì“°ëŠ” íŒ¨í„´ì…ë‹ˆë‹¤. ë°˜ë³µë˜ëŠ” ì½”ë“œë¥¼ ì¤„ì´ê³ , ë°ì´í„°ë¥¼ íš¨ìœ¨ì ìœ¼ë¡œ ê´€ë¦¬í•©ë‹ˆë‹¤.

í•µì‹¬ ê°œë…: @dataclass (ìë™ ìƒì„±ì), @classmethod (íŒ©í† ë¦¬ íŒ¨í„´)

ì¶”ê°€

```python
from dataclasses import dataclass

# @dataclass: __init__, __repr__ ë“±ì„ ìë™ìœ¼ë¡œ ìƒì„±í•´ì¤Œ (ì½”ë“œëŸ‰ ê¸‰ê°)
@dataclass
class Product:
    id: int
    name: str
    price: int

    # @classmethod: ì¸ìŠ¤í„´ìŠ¤ ì—†ì´ í´ë˜ìŠ¤ ìì²´ë¥¼ í†µí•´ í˜¸ì¶œ (ì£¼ë¡œ ê°ì²´ ìƒì„±ìš©)
    @classmethod
    def from_string(cls, data_str):
        # "1,ë…¸íŠ¸ë¶,1500000" ê°™ì€ ë¬¸ìì—´ì„ íŒŒì‹±í•´ì„œ ê°ì²´ ìƒì„±
        p_id, p_name, p_price = data_str.split(',')
        return cls(int(p_id), p_name, int(p_price))

    def apply_discount(self, rate):
        self.price = int(self.price * (1 - rate))

# --- ì‹¤í–‰ ---
# ì¼ë°˜ì ì¸ ìƒì„± ë°©ì‹
p1 = Product(1, "ìŠ¤ë§ˆíŠ¸í°", 1000000)

# í´ë˜ìŠ¤ ë©”ì„œë“œë¥¼ ì´ìš©í•œ ìƒì„± (íŒ©í† ë¦¬ íŒ¨í„´)
csv_data = "2,ê²Œì´ë°ë…¸íŠ¸ë¶,2000000"
p2 = Product.from_string(csv_data)

p2.apply_discount(0.1) # 10% í• ì¸

print(p1)  # Product(id=1, name='ìŠ¤ë§ˆíŠ¸í°', price=1000000)
print(p2)  # Product(id=2, name='ê²Œì´ë°ë…¸íŠ¸ë¶', price=1800000)
```

@dataclassê°€ ì–¼ë§ˆë‚˜ ë§ì€ **'ë°˜ë³µ ë…¸ë™(Boilerplate code)'**ì„ ì¤„ì—¬ì£¼ëŠ”ì§€ í™•ì‹¤í•˜ê²Œ ë¹„êµí•´ ë“œë¦¬ê² ìŠµë‹ˆë‹¤.

ê¸°ëŠ¥ì ìœ¼ë¡œ ì™„ì „íˆ ë™ì¼í•œ ë‘ í´ë˜ìŠ¤(ìƒí’ˆ ì •ë³´ ê´€ë¦¬)ë¥¼ ì‘ì„±í•´ ë³´ê² ìŠµë‹ˆë‹¤.

### 1. ì½”ë“œ ë¹„êµ
(1) @dataclassë¥¼ ì•ˆ ì“´ ê²½ìš° (Old Way)
ìƒì„±ì(__init__), ì¶œë ¥ í˜•íƒœ(__repr__), ë¹„êµ ì—°ì‚°(__eq__)ì„ ëª¨ë‘ ì§ì ‘ êµ¬í˜„í•´ì•¼ í•©ë‹ˆë‹¤.

```python
class ProductOld:
    # 1. ì´ˆê¸°í™” ë©”ì„œë“œ (í•„ìˆ˜)
    def __init__(self, name: str, price: int, category: str):
        self.name = name
        self.price = price
        self.category = category

    # 2. ê°ì²´ë¥¼ ì¶œë ¥í•  ë•Œ ì˜ˆì˜ê²Œ ë³´ì—¬ì£¼ëŠ” ë©”ì„œë“œ (ì—†ìœ¼ë©´ <object at 0x...>ë¡œ ë‚˜ì˜´)
    def __repr__(self):
        return f"ProductOld(name='{self.name}', price={self.price}, category='{self.category}')"

    # 3. ë°ì´í„° ê°’ì´ ê°™ìœ¼ë©´ ê°™ì€ ê°ì²´ë¡œ ì¸ì‹í•˜ê²Œ í•˜ëŠ” ë©”ì„œë“œ (ì—†ìœ¼ë©´ ë©”ëª¨ë¦¬ ì£¼ì†Œë¡œ ë¹„êµ)
    def __eq__(self, other):
        if not isinstance(other, ProductOld):
            return NotImplemented
        return (self.name, self.price, self.category) == (other.name, other.price, other.category)
```

### 1. ê°œë… ì¡ê¸°: ì „ì§ ì‹œìŠ¤í…œ (ìƒì†ê³¼ ì˜¤ë²„ë¼ì´ë”©)
ìƒì† (Inheritance): ì´ˆë³´ì(ë¶€ëª¨)ì˜ ëŠ¥ë ¥(ì´ë¦„, ì²´ë ¥)ì„ ì „ì‚¬(ìì‹)ê°€ ê·¸ëŒ€ë¡œ ë¬¼ë ¤ë°›ëŠ” ê²ƒ. ì½”ë“œë¥¼ ë‹¤ì‹œ ì§¤ í•„ìš”ê°€ ì—†ì–´ì„œ í¸í•©ë‹ˆë‹¤.

ì˜¤ë²„ë¼ì´ë”© (Overriding): ë¬¼ë ¤ë°›ì€ ê¸°ëŠ¥ ì¤‘ ë§ˆìŒì— ì•ˆ ë“œëŠ” ê²ƒì„ ë‚´ ìŠ¤íƒ€ì¼ë¡œ ëœ¯ì–´ê³ ì¹˜ëŠ” ê²ƒ. (ì˜ˆ: 'ì£¼ë¨¹ì§ˆ'ì„ 'ê²€ íœ˜ë“œë¥´ê¸°'ë¡œ ë³€ê²½)

### 2. ì‹¤ìŠµ ì½”ë“œ: ì „ì‚¬ì™€ ë§ˆë²•ì‚¬ ë§Œë“¤ê¸°
ì´ ì½”ë“œëŠ” ì¼ë°˜ í´ë˜ìŠ¤ ë°©ì‹(í‘œì¤€)ìœ¼ë¡œ ì‘ì„±í–ˆìŠµë‹ˆë‹¤. ìƒì†ì˜ íë¦„(super)ì„ ì´í•´í•˜ê¸°ì—ëŠ” ì´ ë°©ì‹ì´ ê°€ì¥ ëª…í™•í•˜ê¸° ë•Œë¬¸ì…ë‹ˆë‹¤.


```python
# -----------------------------------------------
# 1. ë¶€ëª¨ í´ë˜ìŠ¤ (ê¸°ë³¸ ìºë¦­í„°)
# -----------------------------------------------
class Character:
    def __init__(self, name, hp):
        self.name = name
        self.hp = hp
        print(f"ì‘ì• ! {self.name} ìºë¦­í„°ê°€ íƒœì–´ë‚¬ìŠµë‹ˆë‹¤.")

    # ê¸°ë³¸ ê³µê²© (ëª¨ë“  ì§ì—… ê³µí†µ)
    def attack(self):
        print(f"âœŠ {self.name}ì´(ê°€) ì£¼ë¨¹ìœ¼ë¡œ ë•Œë¦½ë‹ˆë‹¤!")

    def move(self):
        print(f"ğŸƒ {self.name}ì´(ê°€) ëšœë²…ëšœë²… ê±¸ì–´ê°‘ë‹ˆë‹¤.")

# -----------------------------------------------
# 2. ìì‹ í´ë˜ìŠ¤ 1: ì „ì‚¬ (Character ìƒì†)
# -----------------------------------------------
class Warrior(Character):
    # attack ë©”ì„œë“œ ì˜¤ë²„ë¼ì´ë”© (ë®ì–´ì“°ê¸°)
    # ë¶€ëª¨ì˜ 'ì£¼ë¨¹ì§ˆ'ì„ ë¬´ì‹œí•˜ê³  ì „ì‚¬ë§Œì˜ ê³µê²©ìœ¼ë¡œ ë°”ê¿ˆ
    def attack(self):
        print(f"âš”ï¸ [ê°•ë ¥í•¨] {self.name}ì´(ê°€) ëŒ€ê²€ì„ íœ˜ë‘ë¦…ë‹ˆë‹¤! ì½°ê´‘!")

# -----------------------------------------------
# 3. ìì‹ í´ë˜ìŠ¤ 2: ë§ˆë²•ì‚¬ (Character ìƒì† + ê¸°ëŠ¥ ì¶”ê°€)
# -----------------------------------------------
class Wizard(Character):
    def __init__(self, name, hp, mp):
        # super(): "ë¶€ëª¨ë‹˜ ê±° ë¨¼ì € í•´ì£¼ì„¸ìš”!"
        # ì´ë¦„ê³¼ ì²´ë ¥ ì„¸íŒ…ì€ ë¶€ëª¨(Character)ì—ê²Œ ë§¡ê¹€
        super().__init__(name, hp)
        self.mp = mp  # ë§ˆë²•ì‚¬ë§Œ ê°€ì§„ 'ë§ˆë‚˜' ì†ì„± ì¶”ê°€

    # attack ë©”ì„œë“œ ì˜¤ë²„ë¼ì´ë”©
    def attack(self):
        print(f"ğŸ”® [ì‹ ë¹„í•¨] {self.name}ì´(ê°€) íŒŒì´ì–´ë³¼ì„ ë˜ì§‘ë‹ˆë‹¤! í™”ë¥´ë¥µ!")

    # ë§ˆë²•ì‚¬ë§Œì˜ ê³ ìœ  ìŠ¤í‚¬ (ë¶€ëª¨ì—ê² ì—†ìŒ)
    def teleport(self):
        print(f"âš¡ {self.name}ì´(ê°€) ìˆœê°„ì´ë™ì„ í–ˆìŠµë‹ˆë‹¤.")

# --- ì‹¤í–‰ ë° í…ŒìŠ¤íŠ¸ ---

print("--- 1. ìºë¦­í„° ìƒì„± ---")
p1 = Character("ê¹€ì‹œë¯¼", 50)   # ê·¸ëƒ¥ ì‚¬ëŒ
p2 = Warrior("ê°•ê°ì°¬", 200)    # ì „ì‚¬
p3 = Wizard("ì´ìˆœì‹ ", 100, 50) # ë§ˆë²•ì‚¬ (ë§ˆë‚˜ 50 ì¶”ê°€)

print("\n--- 2. ê³µê²© í…ŒìŠ¤íŠ¸ (ì˜¤ë²„ë¼ì´ë”© í™•ì¸) ---")
p1.attack()  # ì£¼ë¨¹
p2.attack()  # ëŒ€ê²€ (ì˜¤ë²„ë¼ì´ë”© ë¨)
p3.attack()  # íŒŒì´ì–´ë³¼ (ì˜¤ë²„ë¼ì´ë”© ë¨)

print("\n--- 3. ì´ë™ í…ŒìŠ¤íŠ¸ (ìƒì† í™•ì¸) ---")
# ì „ì‚¬ì™€ ë§ˆë²•ì‚¬ëŠ” moveë¥¼ ë§Œë“¤ì§€ ì•Šì•˜ì§€ë§Œ, ë¶€ëª¨(Character)ì—ê²Œ ë¬¼ë ¤ë°›ì•„ ì‚¬ìš© ê°€ëŠ¥
p2.move()
p3.move()
```
