## 1ë‹¨ê³„: íŒŒì¼ ì“°ê¸° (Write)
íŒŒì¼ì„ ìƒˆë¡œ ìƒì„±í•˜ê±°ë‚˜ ê¸°ì¡´ ë‚´ìš©ì„ ë®ì–´ì”ë‹ˆë‹¤. íŒŒì¼ ì…ì¶œë ¥ ì¤‘ ë°œìƒí•  ìˆ˜ ìˆëŠ” ì‹œìŠ¤í…œ ì—ëŸ¬(IOError)ë¥¼ ì˜ˆì™¸ ì²˜ë¦¬í•©ë‹ˆë‹¤.

```python
# 1ë‹¨ê³„: íŒŒì¼ì— ë‚´ìš© ì“°ê¸° (w ëª¨ë“œ)

file_path = "my_study.txt"

try:
    # 'w' ëª¨ë“œ: íŒŒì¼ì´ ì—†ìœ¼ë©´ ìƒì„±, ìˆìœ¼ë©´ ê¸°ì¡´ ë‚´ìš©ì„ ì‚­ì œí•˜ê³  ìƒˆë¡œ ì”€
    # encoding='utf-8': í•œê¸€ ê¹¨ì§ ë°©ì§€ë¥¼ ìœ„í•´ í•„ìˆ˜ì ìœ¼ë¡œ ì§€ì • ê¶Œì¥
    with open(file_path, "w", encoding="utf-8") as file:
        file.write("1. íŒŒì´ì¬ ê¸°ì´ˆ ê³µë¶€í•˜ê¸°\n")
        file.write("2. íŒŒì¼ ì…ì¶œë ¥ ì˜ˆì œ ì‹¤ìŠµ\n")
        print(f"['{file_path}'] íŒŒì¼ì´ ì„±ê³µì ìœ¼ë¡œ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤.")

except IOError:
    # ë””ìŠ¤í¬ ìš©ëŸ‰ ë¶€ì¡±ì´ë‚˜ ê¶Œí•œ ë¬¸ì œ ë“±ìœ¼ë¡œ íŒŒì¼ì— ì“¸ ìˆ˜ ì—†ì„ ë•Œ ë°œìƒ
    print("íŒŒì¼ì„ ì‘ì„±í•˜ëŠ” ì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤.")
```

## 2ë‹¨ê³„: íŒŒì¼ ì½ê¸° (Read)
ì¡´ì¬í•˜ì§€ ì•ŠëŠ” íŒŒì¼ì„ ì½ìœ¼ë ¤ í•  ë•Œ ë°œìƒí•˜ëŠ” **FileNotFoundError**ë¥¼ ì²˜ë¦¬í•˜ëŠ” ê²ƒì´ í•µì‹¬ì…ë‹ˆë‹¤.

```python
# 2ë‹¨ê³„: íŒŒì¼ ë‚´ìš© ì½ê¸° (r ëª¨ë“œ)

target_file = "my_study.txt"  # ì½ì–´ì˜¬ íŒŒì¼ëª…

try:
    # 'r' ëª¨ë“œ: ì½ê¸° ì „ìš©
    with open(target_file, "r", encoding="utf-8") as file:
        content = file.read()
        print("--- íŒŒì¼ ë‚´ìš© ì‹œì‘ ---")
        print(content)
        print("--- íŒŒì¼ ë‚´ìš© ë ---")

except FileNotFoundError:
    # íŒŒì¼ì´ í•´ë‹¹ ê²½ë¡œì— ì—†ì„ ê²½ìš° ì‹¤í–‰
    print(f"ì˜¤ë¥˜: '{target_file}' íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ê²½ë¡œë¥¼ í™•ì¸í•´ì£¼ì„¸ìš”.")

except Exception as e:
    # ê·¸ ì™¸ ì•Œ ìˆ˜ ì—†ëŠ” ì˜¤ë¥˜ ì²˜ë¦¬
    print(f"ì•Œ ìˆ˜ ì—†ëŠ” ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}")
```

## 3ë‹¨ê³„: ë‚´ìš© ì¶”ê°€í•˜ê¸° (Append) ë° í•œê¸€ ì¸ì½”ë”© ì˜ˆì™¸ ì²˜ë¦¬
ê¸°ì¡´ ë‚´ìš©ì„ ìœ ì§€í•˜ë©´ì„œ ìƒˆë¡œìš´ ë‚´ìš©ì„ ë§ë¶™ì…ë‹ˆë‹¤. ë§Œì•½ ì¸ì½”ë”©ì´ ë§ì§€ ì•ŠëŠ” íŒŒì¼ì„ ì—´ ë•Œ ë°œìƒí•˜ëŠ” UnicodeDecodeError ë“± êµ¬ì²´ì ì¸ ì˜ˆì™¸ ì²˜ë¦¬ë¥¼ ì¶”ê°€í–ˆìŠµë‹ˆë‹¤.

```python
# 3ë‹¨ê³„: ê¸°ì¡´ íŒŒì¼ì— ë‚´ìš© ì¶”ê°€ (a ëª¨ë“œ)

target_file = "my_study.txt"

try:
    # 'a' ëª¨ë“œ: íŒŒì¼ ëì— ë‚´ìš©ì„ ì¶”ê°€ (Append)
    with open(target_file, "a", encoding="utf-8") as file:
        file.write("3. ì˜ˆì™¸ ì²˜ë¦¬ ì½”ë“œ ì‘ì„± ì™„ë£Œ!\n")
        print("ë‚´ìš©ì´ ì„±ê³µì ìœ¼ë¡œ ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤.")

    # ì˜ ì¶”ê°€ë˜ì—ˆëŠ”ì§€ í™•ì¸í•˜ê¸° ìœ„í•´ ë‹¤ì‹œ ì½ì–´ë³´ê¸° (ì¤„ ë‹¨ìœ„ ì½ê¸° ì˜ˆì œ)
    print("\n[ì—…ë°ì´íŠ¸ëœ ë‚´ìš© í™•ì¸]")
    with open(target_file, "r", encoding="utf-8") as file:
        lines = file.readlines() # ë¦¬ìŠ¤íŠ¸ í˜•íƒœë¡œ í•œ ì¤„ì”© ì½ê¸°
        for i, line in enumerate(lines, 1):
            print(f"{i}ë²ˆì§¸ ì¤„: {line.strip()}") # strip()ìœ¼ë¡œ ì¤„ë°”ê¿ˆ ë¬¸ì ì œê±°

except FileNotFoundError:
    print("ìˆ˜ì •í•  íŒŒì¼ì´ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")

except UnicodeDecodeError:
    # ì¸ì½”ë”©(utf-8, cp949 ë“±)ì´ ë§ì§€ ì•Šì•„ ë‚´ìš©ì„ í•´ì„í•  ìˆ˜ ì—†ì„ ë•Œ ë°œìƒ
    print("ì¸ì½”ë”© ì˜¤ë¥˜: íŒŒì¼ ë‚´ìš©ì„ ì½ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
```

## ì‹¤ë¬´ë‚˜ í”„ë¡œì íŠ¸ì—ì„œ ê°€ì¥ ë§ì´ ì“°ì´ëŠ” JSON íŒŒì¼ ì…ì¶œë ¥ ì˜ˆì œ

JSON(JavaScript Object Notation)ì€ íŒŒì´ì¬ì˜ ë”•ì…”ë„ˆë¦¬(Dictionary)ë‚˜ ë¦¬ìŠ¤íŠ¸(List) í˜•íƒœì˜ ë°ì´í„°ë¥¼ ê·¸ëŒ€ë¡œ íŒŒì¼ë¡œ ì €ì¥í•˜ê³  ë¶ˆëŸ¬ì˜¬ ë•Œ ë§¤ìš° ê°•ë ¥í•©ë‹ˆë‹¤. ê²Œì„ ì„¸ì´ë¸Œ ë°ì´í„°, ì„¤ì • íŒŒì¼, í˜¹ì€ ì›¹ í¬ë¡¤ë§ ê²°ê³¼ë¥¼ ì €ì¥í•  ë•Œ í•„ìˆ˜ì ì…ë‹ˆë‹¤.

ğŸ’¡ ì¶”ì²œ ì˜ˆì œ: ê²Œì„ ë°ì´í„°(JSON) ì €ì¥í•˜ê³  ë¶ˆëŸ¬ì˜¤ê¸°
ì´ ì½”ë“œëŠ” ë”•ì…”ë„ˆë¦¬ ë°ì´í„°ë¥¼ game_data.json íŒŒì¼ë¡œ ì €ì¥í•˜ê³ , ë‹¤ì‹œ ë¶ˆëŸ¬ì™€ì„œ íŒŒì´ì¬ ë”•ì…”ë„ˆë¦¬ë¡œ ë³€í™˜í•˜ëŠ” ê³¼ì •ì„ ë³´ì—¬ì¤ë‹ˆë‹¤.

### ì´ ì½”ë“œê°€ 'ì¢‹ì€ ì˜ˆì œ'ì¸ ì´ìœ  (í•µì‹¬ í¬ì¸íŠ¸)
import json: íŒŒì´ì¬ ê¸°ë³¸ ë‚´ì¥ ëª¨ë“ˆì´ë¼ ë³„ë„ ì„¤ì¹˜ê°€ í•„ìš” ì—†ìŠµë‹ˆë‹¤.

ensure_ascii=False: ì´ ì˜µì…˜ì´ ì—†ìœ¼ë©´ í•œê¸€ì´ \uDO... ê°™ì€ ìœ ë‹ˆì½”ë“œ ë¬¸ìë¡œ ì €ì¥ë˜ì–´ ì‚¬ëŒì´ ì½ê¸° ì–´ë µìŠµë‹ˆë‹¤. í•œê¸€ ë°ì´í„°ë¥¼ ë‹¤ë£° ë•Œ í•„ìˆ˜ì…ë‹ˆë‹¤.

indent=4: ì €ì¥ëœ íŒŒì¼ì„ ë©”ëª¨ì¥ìœ¼ë¡œ ì—´ì—ˆì„ ë•Œ, ì¤„ë°”ê¿ˆê³¼ ë“¤ì—¬ì“°ê¸°ê°€ ë˜ì–´ ìˆì–´ ê°€ë…ì„±ì´ ë§¤ìš° ì¢‹ì•„ì§‘ë‹ˆë‹¤.

json.JSONDecodeError: í…ìŠ¤íŠ¸ íŒŒì¼ê³¼ ë‹¬ë¦¬, JSONì€ ê´„í˜¸({})ë‚˜ ì‰¼í‘œ(,) í•˜ë‚˜ë§Œ ë¹ ì ¸ë„ ì—ëŸ¬ê°€ ë‚©ë‹ˆë‹¤. ì´ ì „ìš© ì˜ˆì™¸ ì²˜ë¦¬ë¥¼ í†µí•´ íŒŒì¼ í˜•ì‹ì´ ê¹¨ì¡ŒëŠ”ì§€ ì •í™•íˆ ì¡ì•„ë‚¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ì´ ì½”ë“œë¥¼ ì‹¤í–‰í•´ ë³´ì‹œë©´ í´ë”ì— game_data.json íŒŒì¼ì´ ì˜ˆì˜ê²Œ ìƒì„±ëœ ê²ƒì„ í™•ì¸í•˜ì‹¤ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
import json # JSON ì²˜ë¦¬ë¥¼ ìœ„í•œ ë‚´ì¥ ëª¨ë“ˆ

file_path = "game_data.json"

# 1. ì €ì¥í•  ë°ì´í„° (ë”•ì…”ë„ˆë¦¬ ë¦¬ìŠ¤íŠ¸ í˜•íƒœ)
player_data = {
    "nickname": "íŒŒì´ì¬ê³ ìˆ˜",
    "level": 15,
    "items": ["í™©ê¸ˆê²€", "íšŒë³µí¬ì…˜", "ì§€ë„"],
    "is_active": True
}

# --- [ì“°ê¸°] JSON íŒŒì¼ë¡œ ì €ì¥í•˜ê¸° ---
try:
    with open(file_path, "w", encoding="utf-8") as file:
        # dump: íŒŒì´ì¬ ë°ì´í„°ë¥¼ JSON í˜•ì‹ìœ¼ë¡œ ë³€í™˜í•´ íŒŒì¼ì— ì”€
        # ensure_ascii=False: í•œê¸€ì´ ê¹¨ì§€ì§€ ì•Šê³  ê·¸ëŒ€ë¡œ ì €ì¥ë¨
        # indent=4: ë“¤ì—¬ì“°ê¸°ë¥¼ í•´ì„œ ì‚¬ëŒì´ ë³´ê¸° ì¢‹ê²Œ ì •ë ¬í•¨
        json.dump(player_data, file, ensure_ascii=False, indent=4)

    print(f"ë°ì´í„°ê°€ '{file_path}'ì— ì„±ê³µì ìœ¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")

except Exception as e:
    print(f"ì €ì¥ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")


print("\n" + "="*30 + "\n")


# --- [ì½ê¸°] JSON íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸° ---
try:
    with open(file_path, "r", encoding="utf-8") as file:
        # load: íŒŒì¼ì˜ JSON ë‚´ìš©ì„ ì½ì–´ì„œ íŒŒì´ì¬ ë”•ì…”ë„ˆë¦¬ë¡œ ë³€í™˜
        loaded_data = json.load(file)

        print(f"ë¶ˆëŸ¬ì˜¨ ë°ì´í„° íƒ€ì…: {type(loaded_data)}") # <class 'dict'> í™•ì¸
        print(f"í”Œë ˆì´ì–´ ì´ë¦„: {loaded_data['nickname']}")
        print(f"ë³´ìœ  ì•„ì´í…œ: {loaded_data['items']}")

except FileNotFoundError:
    print("íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ê²½ë¡œë¥¼ í™•ì¸í•˜ì„¸ìš”.")

except json.JSONDecodeError:
    # íŒŒì¼ì€ ìˆì§€ë§Œ ë‚´ìš©ì´ ì˜¬ë°”ë¥¸ JSON í˜•ì‹ì´ ì•„ë‹ ë•Œ ë°œìƒ
    print("JSON í˜•ì‹ì´ ì˜ëª»ë˜ì–´ ë°ì´í„°ë¥¼ í•´ì„í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

except Exception as e:
    print(f"ì•Œ ìˆ˜ ì—†ëŠ” ì˜¤ë¥˜ ë°œìƒ: {e}")
```

## í´ë” ìë™ ìƒì„± + CSV íŒŒì¼ ë‹¤ë£¨ê¸° í†µí•© ì˜ˆì œ
ì´ ì½”ë“œëŠ” data_filesë¼ëŠ” í´ë”ê°€ ì—†ìœ¼ë©´ ì•Œì•„ì„œ ë§Œë“¤ê³ , ê·¸ ì•ˆì— my_stocks.csv íŒŒì¼ì„ ìƒì„±í•©ë‹ˆë‹¤.

ì•„ë˜ ì½”ë“œëŠ” ê´€ì‹¬ì‚¬(ì£¼ì‹, ë°ì´í„° ë¶„ì„)ì— ë§ì¶°, 'ê´€ì‹¬ ì¢…ëª© ë¦¬ìŠ¤íŠ¸'ë¥¼ ê´€ë¦¬í•˜ëŠ” ì‹¤ì „ ì½”ë“œë¡œ ë§Œë“¤ì–´ ë³´ì•˜ìŠµë‹ˆë‹¤.

### ì´ ì½”ë“œì˜ 'ê¿€íŒ' í¬ì¸íŠ¸
os.makedirs(..., exist_ok=True):

ê³¼ê±°ì—ëŠ” if not os.path.exists... ì²˜ëŸ¼ if ë¬¸ì„ ì¨ì„œ ê²€ì‚¬í–ˆì§€ë§Œ, ì´ì œëŠ” ì´ ì˜µì…˜ í•˜ë‚˜ë©´ "ì—†ìœ¼ë©´ ë§Œë“¤ê³ , ìˆìœ¼ë©´ ê·¸ëƒ¥ ì¨!"ê°€ ë˜ì–´ ì½”ë“œê°€ í›¨ì”¬ ê¹”ë”í•´ì§‘ë‹ˆë‹¤.

encoding='utf-8-sig':

ê·¸ëƒ¥ utf-8ë¡œ ì €ì¥í•˜ë©´ ì—‘ì…€(Excel)ì—ì„œ ì—´ì—ˆì„ ë•Œ í•œê¸€ì´ ì™¸ê³„ì–´ì²˜ëŸ¼ ê¹¨ì§ˆ ìˆ˜ ìˆìŠµë‹ˆë‹¤. **utf-8-sig**ë¥¼ ì“°ë©´ ì—‘ì…€ì—ì„œë„ í•œê¸€ì´ ì™„ë²½í•˜ê²Œ ë³´ì…ë‹ˆë‹¤. ë°ì´í„° ë¶„ì„ ë°ì´í„°ë¥¼ ì¹œêµ¬ë‚˜ ì„ ìƒë‹˜ê»˜ ë³´ë‚¼ ë•Œ í•„ìˆ˜ì…ë‹ˆë‹¤.

newline='':

íŒŒì´ì¬ 3ì—ì„œ CSVë¥¼ ë‹¤ë£° ë•Œ ì´ ì˜µì…˜ì„ ì•ˆ ì“°ë©´, ì—‘ì…€ì—ì„œ ì—´ì—ˆì„ ë•Œ ì¤„ ì‚¬ì´ì— ë¹ˆ ì¤„ì´ í•˜ë‚˜ì”© ë” ìƒê¸°ëŠ” í˜„ìƒì´ ë°œìƒí•©ë‹ˆë‹¤. ì´ë¥¼ ë§‰ê¸° ìœ„í•´ ê¼­ ë„£ì–´ì£¼ì„¸ìš”.

```python
import os   # í´ë”(ë””ë ‰í† ë¦¬) ê´€ë¦¬ë¥¼ ìœ„í•œ ëª¨ë“ˆ
import csv  # CSV íŒŒì¼ ì²˜ë¦¬ë¥¼ ìœ„í•œ ëª¨ë“ˆ

# 1. ì €ì¥í•  ê²½ë¡œì™€ íŒŒì¼ëª… ì„¤ì •
folder_name = "data_files"
file_name = "my_stocks.csv"
full_path = os.path.join(folder_name, file_name) # ìš´ì˜ì²´ì œì— ë§ê²Œ ê²½ë¡œ í•©ì¹˜ê¸° (data_files/my_stocks.csv)

# --- [1ë‹¨ê³„] í´ë” ìë™ ìƒì„±í•˜ê¸° ---
try:
    # exist_ok=True: í´ë”ê°€ ì´ë¯¸ ìˆì–´ë„ ì—ëŸ¬ë¥¼ ë‚´ì§€ ì•Šê³  ë„˜ì–´ê° (ê°€ì¥ ì•ˆì „í•œ ë°©ë²•)
    os.makedirs(folder_name, exist_ok=True)
    print(f"í´ë” í™•ì¸ ì™„ë£Œ: '{folder_name}'")
except OSError as e:
    print(f"í´ë” ìƒì„± ì‹¤íŒ¨: {e}")

print("-" * 30)

# --- [2ë‹¨ê³„] CSV íŒŒì¼ ì“°ê¸° (Write) ---
# ì €ì¥í•  ë°ì´í„° (í—¤ë” + ë‚´ìš©)
header = ["ì¢…ëª©ëª…", "í˜„ì¬ê°€", "ë³´ìœ ìˆ˜ëŸ‰"]
data = [
    ["ì‚¼ì„±ì „ì", 75000, 10],
    ["Apple", 190, 5],
    ["Hanwha Aerospace", 150000, 3]
]

try:
    # encoding='utf-8-sig': ì—‘ì…€ì—ì„œ í•œê¸€ì´ ê¹¨ì§€ì§€ ì•Šê²Œ í•˜ëŠ” 'ë§ˆë²•ì˜ ì˜µì…˜'ì…ë‹ˆë‹¤.
    # newline='': ìœˆë„ìš°ì—ì„œ ì¤„ë°”ê¿ˆì´ ë‘ ë²ˆ ì¼ì–´ë‚˜ëŠ” ê²ƒì„ ë°©ì§€í•©ë‹ˆë‹¤.
    with open(full_path, "w", encoding="utf-8-sig", newline='') as file:
        writer = csv.writer(file)

        writer.writerow(header) # ì²« ì¤„(ì œëª©) ì“°ê¸°
        writer.writerows(data)  # ë‚˜ë¨¸ì§€ ë°ì´í„° í•œêº¼ë²ˆì— ì“°ê¸°

    print(f"âœ… CSV íŒŒì¼ ì €ì¥ ì™„ë£Œ: {full_path}")

except Exception as e:
    print(f"íŒŒì¼ ì“°ê¸° ì˜¤ë¥˜: {e}")

print("-" * 30)

# --- [3ë‹¨ê³„] CSV íŒŒì¼ ì½ê¸° (Read) ---
print(f"[{file_name} ë‚´ìš© ì½ì–´ì˜¤ê¸°]")

try:
    with open(full_path, "r", encoding="utf-8-sig") as file:
        reader = csv.reader(file)

        # í•œ ì¤„ì”© ê°€ì ¸ì™€ì„œ ì¶œë ¥
        for row in reader:
            print(row) # ê° ì¤„ì€ ë¦¬ìŠ¤íŠ¸ í˜•íƒœë¡œ ì¶œë ¥ë©ë‹ˆë‹¤ ['ì¢…ëª©ëª…', 'í˜„ì¬ê°€', ...]

except FileNotFoundError:
    print("íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
```
