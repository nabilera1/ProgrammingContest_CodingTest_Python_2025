```python
import random

# 1. 1ë¶€í„° 50 ì‚¬ì´ì˜ ëœë¤ ìˆ«ìë¥¼ ë½‘ìŠµë‹ˆë‹¤.
number = random.randint(1, 50)

# 2. ì‹œë„ íšŸìˆ˜ë¥¼ ì…€ ë³€ìˆ˜ë¥¼ ë§Œë“¤ê³  0ìœ¼ë¡œ ì´ˆê¸°í™”í•©ë‹ˆë‹¤.
count = 0

print("ì»´í“¨í„°: ìˆ«ìë¥¼ ê²°ì •í–ˆìŠµë‹ˆë‹¤. ë§í ë•Œê¹Œì§€ ê³„ì† ë„ì „í•´ë³´ì„¸ìš”!")

# 3. ì •ë‹µì„ ë§í ë•Œê¹Œì§€ ë¬´í•œ ë°˜ë³µí•©ë‹ˆë‹¤.
while True:
    count += 1  # ì‹œë„ë¥¼ í•  ë•Œë§ˆë‹¤ íšŸìˆ˜ë¥¼ 1ì”© ì¦ê°€ì‹œí‚µë‹ˆë‹¤.

    print(f"\n[{count}ë²ˆì§¸ ë„ì „]")
    guess = int(input("ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš” (1~50): "))

    if guess == number:
        # ì •ë‹µì„ ë§íˆë©´ ì¶•í•˜ ë©”ì‹œì§€ì™€ ì´ ì‹œë„ íšŸìˆ˜ë¥¼ ë³´ì—¬ì¤ë‹ˆë‹¤.
        print("ğŸ‰ ì •ë‹µì…ë‹ˆë‹¤!")
        print(f"ì¶•í•˜í•©ë‹ˆë‹¤! ì´ {count}ë²ˆ ë§Œì— ë§íˆì…¨ìŠµë‹ˆë‹¤.")
        break  # ê²Œì„ì„ ì¢…ë£Œ(ë°˜ë³µë¬¸ íƒˆì¶œ)í•©ë‹ˆë‹¤.

    elif guess < number:
        print("Up! (ë” í° ìˆ˜ì…ë‹ˆë‹¤)")

    else:
        print("Down! (ë” ì‘ì€ ìˆ˜ì…ë‹ˆë‹¤)")
```
