## [Level 1] íŒŒì´ì¬ ì œì–´ë¬¸(if, while)ì„ ë§‰ ë—€ í•™ìƒìš©

í•™ìŠµ ëª©í‘œ: random ëª¨ë“ˆ ì‚¬ìš©ë²•, input í˜•ë³€í™˜, while ë°˜ë³µë¬¸ê³¼ if ì¡°ê±´ë¬¸ì˜ íë¦„ ì´í•´.

```python
import random  # ëœë¤í•œ ìˆ«ìë¥¼ ë½‘ê¸° ìœ„í•´ ëª¨ë“ˆì„ ê°€ì ¸ì˜µë‹ˆë‹¤.

# 1ë¶€í„° 100 ì‚¬ì´ì˜ ë‚œìˆ˜(ì •ë‹µ)ë¥¼ í•˜ë‚˜ ìƒì„±í•©ë‹ˆë‹¤.
answer = random.randint(1, 100)
count = 0  # ì‹œë„ íšŸìˆ˜ë¥¼ ì„¸ê¸° ìœ„í•œ ë³€ìˆ˜

print("=== [Level 1] ìˆ«ì ë§ì¶”ê¸° ê²Œì„ (1~100) ===")

while True:  # ì •ë‹µì„ ë§ì¶œ ë•Œê¹Œì§€ ë¬´í•œ ë°˜ë³µí•©ë‹ˆë‹¤.
    # input()ì€ ë¬¸ìì—´ë¡œ ì…ë ¥ì„ ë°›ìœ¼ë¯€ë¡œ int()ë¥¼ ì‚¬ìš©í•´ ì •ìˆ˜ë¡œ ë°”ê¿”ì¤ë‹ˆë‹¤.
    guess = int(input("ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))
    count = count + 1  # ì‹œë„ íšŸìˆ˜ 1 ì¦ê°€

    if guess == answer:  # ì •ë‹µì„ ë§ì¶˜ ê²½ìš°
        print("ì •ë‹µì…ë‹ˆë‹¤!")
        print(count, "ë²ˆ ë§Œì— ë§ì·„ë„¤ìš”.")
        break  # ë°˜ë³µë¬¸ì„ íƒˆì¶œí•˜ì—¬ ê²Œì„ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.

    elif guess > answer:  # ì…ë ¥í•œ ê°’ì´ ì •ë‹µë³´ë‹¤ í´ ê²½ìš°
        print("ë” ì‘ì€ ìˆ«ìì…ë‹ˆë‹¤ (Down)")

    else:  # ì…ë ¥í•œ ê°’ì´ ì •ë‹µë³´ë‹¤ ì‘ì„ ê²½ìš° (guess < answer)
        print("ë” í° ìˆ«ìì…ë‹ˆë‹¤ (Up)")
```

## [Level 2] ë¦¬ìŠ¤íŠ¸ì™€ ë°˜ë³µë¬¸(for)ì„ ë°°ìš´ í•™ìƒìš©

í•™ìŠµ ëª©í‘œ: list.appendë¡œ ê¸°ë¡ ì €ì¥í•˜ê¸°, f-string í¬ë§·íŒ…, ê²Œì„ íšŸìˆ˜ ì œí•œ(Game Over) êµ¬í˜„.

```python
import random

answer = random.randint(1, 100)
history = []  # ì‚¬ìš©ìê°€ ì…ë ¥í•œ ìˆ«ìë“¤ì„ ê¸°ë¡í•  ë¹ˆ ë¦¬ìŠ¤íŠ¸
max_try = 10  # ìµœëŒ€ ì‹œë„ íšŸìˆ˜ ì œí•œ

print(f"=== [Level 2] ê¸°íšŒëŠ” {max_try}ë²ˆ! ê¸°ë¡ì´ ë‚¨ëŠ” ê²Œì„ ===")

# ìµœëŒ€ íšŸìˆ˜(max_try)ë§Œí¼ë§Œ ë°˜ë³µí•˜ê±°ë‚˜, ì •ë‹µì„ ë§ì¶”ë©´ ì¢…ë£Œí•©ë‹ˆë‹¤.
while len(history) < max_try:
    guess = int(input(f"{len(history) + 1}ë²ˆì§¸ ì‹œë„: ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))

    # ì…ë ¥í•œ ìˆ«ìë¥¼ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€(ê¸°ë¡)í•©ë‹ˆë‹¤.
    history.append(guess)

    if guess == answer:
        print(f"ì¶•í•˜í•©ë‹ˆë‹¤! {len(history)}ë²ˆ ë§Œì— ì •ë‹µ {answer}ì„(ë¥¼) ë§ì·„ìŠµë‹ˆë‹¤.")
        break  # ì •ë‹µ ì‹œ ì¦‰ì‹œ ì¢…ë£Œ
    elif guess > answer:
        print("Down! ë” ì‘ì€ ìˆ«ìì…ë‹ˆë‹¤.")
    else:
        print("Up! ë” í° ìˆ«ìì…ë‹ˆë‹¤.")

# ë°˜ë³µë¬¸ì´ ëë‚œ í›„, ì •ë‹µì„ ëª» ë§ì·„ëŠ”ì§€ í™•ì¸
if answer not in history: # ë¦¬ìŠ¤íŠ¸ ì•ˆì— ì •ë‹µì´ ì—†ë‹¤ë©´ ì‹¤íŒ¨í•œ ê²ƒ
    print(f"\n[Game Over] ê¸°íšŒë¥¼ ëª¨ë‘ ì†Œì§„í–ˆìŠµë‹ˆë‹¤. ì •ë‹µì€ {answer}ì˜€ìŠµë‹ˆë‹¤.")

# ê²Œì„ì´ ëë‚œ í›„ ì‚¬ìš©ìì˜ ê¸°ë¡ì„ ë³´ì—¬ì¤ë‹ˆë‹¤.
print("\n=== í”Œë ˆì´ ê¸°ë¡ ===")
# ë¦¬ìŠ¤íŠ¸ì— ì €ì¥ëœ ê¸°ë¡ì„ í•˜ë‚˜ì”© êº¼ë‚´ì„œ ì¶œë ¥
for h in history:
    print(h, end=" -> ")
print("ë")
```

## [Level 3] íŒŒì´ì¬ ê¸°ì´ˆ ì±… í•œ ê¶Œì„ ë—€ í•™ìƒìš©

í•™ìŠµ ëª©í‘œ: def í•¨ìˆ˜ë¥¼ ì´ìš©í•œ ëª¨ë“ˆí™”, try-except ì˜ˆì™¸ ì²˜ë¦¬(ì—ëŸ¬ ë°©ì§€), main í•¨ìˆ˜ êµ¬ì¡°, ì½”ë“œì˜ ì¬ì‚¬ìš©ì„±.

```python
import random

def get_valid_input():
    """
    ì‚¬ìš©ìë¡œë¶€í„° ì •ìˆ˜ë¥¼ ì•ˆì „í•˜ê²Œ ì…ë ¥ë°›ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤.
    ìˆ«ìê°€ ì•„ë‹Œ ë¬¸ìë¥¼ ì…ë ¥í–ˆì„ ë•Œì˜ ì—ëŸ¬(ValueError)ë¥¼ ì²˜ë¦¬í•©ë‹ˆë‹¤.
    """
    while True:
        try:
            user_input = input("ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš” (1~100): ")
            return int(user_input)  # ì •ìˆ˜ë¡œ ë³€í™˜í•˜ì—¬ ë°˜í™˜
        except ValueError:
            # ìˆ«ìê°€ ì•„ë‹Œ ê°’ì„ ì…ë ¥í–ˆì„ ë•Œ í”„ë¡œê·¸ë¨ì´ ë©ˆì¶”ì§€ ì•Šê³  ê²½ê³  ë©”ì‹œì§€ë¥¼ ë„ì›€
            print("[ì˜¤ë¥˜] ìœ íš¨í•œ ìˆ«ìê°€ ì•„ë‹™ë‹ˆë‹¤. ë‹¤ì‹œ ì…ë ¥í•´ì£¼ì„¸ìš”.")

def play_game():
    """
    ê²Œì„ì˜ í•µì‹¬ ë¡œì§ì„ ë‹´ë‹¹í•˜ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤.
    """
    answer = random.randint(1, 100)
    attempts = 0
    print("\n=== [Level 3] ì˜¤ë¥˜ ì—†ëŠ” ê²¬ê³ í•œ ê²Œì„ ì‹œì‘ ===")

    while True:
        attempts += 1
        guess = get_valid_input()  # ìœ„ì—ì„œ ë§Œë“  í•¨ìˆ˜ë¥¼ í˜¸ì¶œí•´ ì…ë ¥ì„ ë°›ìŒ

        if guess < 1 or guess > 100:
            print("ë²”ìœ„ë¥¼ ë²—ì–´ë‚¬ìŠµë‹ˆë‹¤. 1ë¶€í„° 100 ì‚¬ì´ì˜ ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”.")
            continue # ì•„ë˜ ì½”ë“œë¥¼ ì‹¤í–‰í•˜ì§€ ì•Šê³  ë‹¤ì‹œ ë°˜ë³µë¬¸ ì²˜ìŒìœ¼ë¡œ ëŒì•„ê°

        if guess == answer:
            print(f"ğŸ‰ ì •ë‹µì…ë‹ˆë‹¤! {attempts}ë²ˆ ë§Œì— ì„±ê³µí•˜ì…¨ìŠµë‹ˆë‹¤.")
            return True # ê²Œì„ ìŠ¹ë¦¬ ë°˜í™˜
        elif guess > answer:
            print(f"{guess}ë³´ë‹¤ ì‘ìŠµë‹ˆë‹¤. (Down)")
        else:
            print(f"{guess}ë³´ë‹¤ í½ë‹ˆë‹¤. (Up)")

def main():
    """
    í”„ë¡œê·¸ë¨ì˜ ì‹œì‘ì ì…ë‹ˆë‹¤. ì¬ê²½ê¸° ì—¬ë¶€ë¥¼ ë¬»ìŠµë‹ˆë‹¤.
    """
    while True:
        play_game() # ê²Œì„ í•œ íŒ ì‹¤í–‰

        # ê²Œì„ ì¢…ë£Œ í›„ ì¬ì‹œì‘ ì—¬ë¶€ í™•ì¸
        replay = input("ê²Œì„ì„ ë‹¤ì‹œ í•˜ì‹œê² ìŠµë‹ˆê¹Œ? (y/n): ").lower()
        if replay != 'y':
            print("ê²Œì„ì„ ì¢…ë£Œí•©ë‹ˆë‹¤. ì¦ê±°ìš´ í•˜ë£¨ ë˜ì„¸ìš”!")
            break

# ì´ íŒŒì¼ì´ ì§ì ‘ ì‹¤í–‰ë  ë•Œë§Œ main() í•¨ìˆ˜ë¥¼ ì‹¤í–‰í•©ë‹ˆë‹¤.
if __name__ == "__main__":
    main()
```
