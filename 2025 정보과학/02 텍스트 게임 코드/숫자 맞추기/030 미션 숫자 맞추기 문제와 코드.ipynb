isdigit()은 문자열이 '모두 숫자'로만 이루어져 있는지 확인할 때 사용합니다. (음수 기호 -나 소수점 .이 있으면 False가 나옵니다.)

```python
# text = "12345"  # 문자열이지만 내용이 모두 숫자
text = input('숫자를 입력하세요 : ')
# text = int(input('숫자를 입력하세요 : ')) # AttributeError: 'int' object has no attribute 'isdigit'
if text.isdigit():
    print("이것은 숫자입니다.")
    # 숫자로 확인되었으니 정수형으로 변환 가능
    number = int(text)
else:
    print("문자가 섞여 있습니다.")
```

참고

"-10".isdigit() → False (마이너스 기호 때문)

"3.14".isdigit() → False (소수점 때문)

```python
age = "20살"  # 숫자와 한글이 섞여 있음

if age.isdigit():
    print("나이가 정상적으로 입력되었습니다.")
else:
    print("숫자만 입력해주세요.")  # 이 부분이 실행됨
```

```python
'''
isdigit()은 - 기호나 . 소수점을 '숫자'가 아닌 '문자'로 인식하기 때문에, 음수나 소수를 처리할 때는 try-except 구문을 쓰는 것이 가장 확실한 방법입니다.

변환을 시도(try) 해보고, 안 되면 에러를 잡아내는(except) 방식입니다.

1. 음수(정수) 처리하기
int()로 변환을 시도합니다. -10 같은 문자열도 숫자로 잘 변환됩니다.

'''

text = "-50"

try:
    # 일단 숫자로 변환을 시도합니다
    number = int(text)
    print(f"변환 성공: {number}")

except ValueError:
    # 변환에 실패하면(숫자가 아니면) 이쪽으로 옵니다
    print("숫자가 아닙니다.")
```

```python
'''
2. 소수(실수) 처리하기
float()로 변환을 시도합니다. 3.14 같은 소수점 문자열을 처리할 수 있습니다.
'''

text = "3.14"

try:
    # 실수(소수)로 변환을 시도합니다
    number = float(text)
    print(f"변환 성공: {number}")

except ValueError:
    # "3.14.15" 처럼 점이 두 개거나, 문자가 섞이면 실패합니다
    print("소수가 아닙니다.")

'''
[핵심 차이점]
isdigit(): "글자 모양이 숫자인가?"만 검사 (단순함)

try-except: "실제로 숫자 계산이 가능한가?"를 검사 (강력함, 음수/소수 가능)
'''
```

# 숫자 추측 미니게임

```python
"""
문제) 숫자 추측 미니게임

[문제] 다음 조건을 만족하는 숫자 추측 프로그램을 작성하시오.
1~20 사이 무작위 정답을 만든다.
사용자는 총 5번 추측할 수 있다.
숫자가 아니면 “숫자!”를 출력하고 기회를 소모하지 않는다.
정답이면 즉시 종료, 아니면 UP/DOWN 힌트를 준다.
기회 소진 시 정답을 출력한다.
"""

import random
lo, hi, tries = 1, 20, 5
secret = random.randint(lo, hi)
print(f"{lo}~{hi} 숫자 맞히기! {tries}번 기회")
while tries > 0:
    s = input("추측: ").strip()
    if not s.isdigit():
        print("숫자!"); continue
    g = int(s)
    if g == secret:
        print("정답!")
        break
    print("UP" if g < secret else "DOWN")
    tries -= 1
if tries == 0:
    print("실패.. 정답:", secret)
```
