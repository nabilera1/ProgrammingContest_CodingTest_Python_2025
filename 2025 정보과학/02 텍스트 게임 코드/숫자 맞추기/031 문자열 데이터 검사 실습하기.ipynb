```python
1. isnumeric(): 숫자처럼 생긴 건 다 인정해주는 함수

isdigit()보다 범위가 더 넓습니다. 단순히 0~9 숫자뿐만 아니라, 분수(½)나 로마 숫자(Ⅱ) 같은 특수 문자도 숫자로 인식합니다.
```

```python
text1 = "123"
text2 = "½"  # 분수 특수문자

print(text1.isnumeric())  # True
print(text2.isnumeric())  # True (isdigit은 여기서 False가 나옴)
```

```python
2. isalpha(): 글자로만 되어 있는지 확인 (숫자 금지)
문자열이 한글이나 알파벳으로만 구성되어 있는지 확인합니다. 공백, 숫자, 특수문자가 섞여 있으면 False가 나옵니다. 이름을 입력받을 때 유용합니다.
```

```python
name = "홍길동"
id_card = "홍길동1"

print(name.isalpha())     # True
print(id_card.isalpha())  # False (숫자 1이 섞여 있어서)
```

```python
3. isalnum(): 글자와 숫자가 섞여 있어도 OK
**알파벳(Alpha) + 숫자(Numeric)**의 줄임말입니다.
특수문자나 공백 없이, 오직 글자와 숫자로만 이루어져 있는지 확인할 때 씁니다.
아이디(ID)나 닉네임 규칙을 검사할 때 가장 많이 씁니다.
```

```python
user_id = "user123"
password = "user123!"

print(user_id.isalnum())   # True (영어+숫자 조합이라 OK)
print(password.isalnum())  # False (느낌표 ! 특수문자 때문에 탈락)
```
