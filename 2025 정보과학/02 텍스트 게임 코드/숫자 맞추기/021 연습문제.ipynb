## ë‹¤ìŒ ì½”ë“œì—ì„œ ë°œìƒí•˜ëŠ” ë¬¸ì œê°€ ë¬´ì—‡ì¸ì§€ ë§í•˜ê³ , í•´ê²°í•˜ì‹œì˜¤


```python
mport random

history = []
num = 0
while True:
    number = random.randint(1,50)
    print("\n"+ "=" *40)
    print("ìƒˆë¡œìš´ ê²Œì„ ì‹œì‘! ê¸°íšŒëŠ” 10ë²ˆ")
    print("*" * 40)

    is_success = False
    if is_success == False:
        for i in range(10): #10ë²ˆì˜ ê¸°íšŒ
            print(f'\n{i+1}ë²ˆì§¸ ì‹œë„')
            guess = int(input('ìˆ«ì ì…ë ¥(1~50) : '))
            history.append(guess)
            if guess == number:
                print(f'ì¶•í•˜í•©ë‹ˆë‹¤! {num}ë²ˆ ë§Œì— ì •ë‹µ:{number}ë¥¼ ë§ì·„ìŠµë‹ˆë‹¤.')
                print("=== í”Œë ˆì´ ê¸°ë¡ ===")
                for i in range(0,len(history)):
                    print(history[i],end="->")
                is_success = True #í”Œë˜ìŠ¤
                print("ë")
                break

            elif guess < number :
                print('up,ë” í°ìˆ«ì ì…ë‹ˆë‹¤.')
                num += 1
            else :
                print('Down, ë” ì‘ì€ìˆ˜ ì…ë‹ˆë‹¤.')
                num += 1

        #rê²Œì„ ê²°ê³¼ ì²˜ë¦¬
        if is_success == False:
            print(f"ê¸°íšŒë¥¼ ëª¨ë‘ ì‚¬ìš©!, ì •ë‹µì€ {number}")
            retry = input('\në‹¤ì‹œ ì‹œë„? (y/n) :')

            # if retry != 'y' and retry != 'Y':
            #     print('Game over!')
            #     break
            if not retry.lower().startswith('y'):
                print('Game over!')
                break
```

```python
# ì •ë‹µì„ ë§ì·„ì„ ë•Œ(ì„±ê³µí–ˆì„ ë•Œ) ê²Œì„ì´ ëë‚˜ê±°ë‚˜ ë‹¤ì‹œ í• ì§€ ë¬»ì§€ ì•Šê³ , ë¬´ì¡°ê±´ ë°”ë¡œ ìƒˆ ê²Œì„ì´ ì‹œì‘ë˜ëŠ” ë¬¸ì œê°€ ìˆìŠµë‹ˆë‹¤.
#
# ğŸ” ì›ì¸ ë¶„ì„ (ë²”ì¸ì€ 'ë“¤ì—¬ì“°ê¸°'ì™€ 'ìœ„ì¹˜')
# í˜„ì¬ ì½”ë“œ êµ¬ì¡°ë¥¼ ë³´ë©´ "ë‹¤ì‹œ ì‹œë„í•˜ì‹œê² ìŠµë‹ˆê¹Œ?"ë¼ê³  ë¬»ëŠ” ë¶€ë¶„ì´ if is_success == False: ë¸”ë¡ ì•ˆì— ë“¤ì–´ìˆìŠµë‹ˆë‹¤.
#
# ì‹¤íŒ¨í–ˆì„ ë•Œ (is_success == False): ì´ ì•ˆìœ¼ë¡œ ë“¤ì–´ì™€ì„œ "ë‹¤ì‹œ ì‹œë„?"ë¥¼ ë¬¼ì–´ë´…ë‹ˆë‹¤. (ì •ìƒ)
#
# ì„±ê³µí–ˆì„ ë•Œ (is_success == True): is_success == False ì¡°ê±´ì´ ë§ì§€ ì•Šìœ¼ë¯€ë¡œ ì´ ë¸”ë¡ì„ í†µì§¸ë¡œ ê±´ë„ˆëœë‹ˆë‹¤.
#
# ê²°êµ­ while True:ì˜ ì²˜ìŒìœ¼ë¡œ ëŒì•„ê°€ì„œ ë¬»ì§€ë„ ë”°ì§€ì§€ë„ ì•Šê³  ê²Œì„ì„ ê°•ì œë¡œ ë‹¤ì‹œ ì‹œì‘í•©ë‹ˆë‹¤.
#
# ğŸ›  í•´ê²° ë°©ë²•
# "ë‹¤ì‹œ ì‹œë„?"ë¥¼ ë¬»ëŠ” ì½”ë“œë¥¼ ifë¬¸ ë°–ìœ¼ë¡œ êº¼ë‚´ì•¼ í•©ë‹ˆë‹¤. ê·¸ë˜ì•¼ ì´ê¸°ë“  ì§€ë“  ìƒê´€ì—†ì´ í•­ìƒ ê²Œì„ì´ ëë‚œ í›„ ë¬¼ì–´ë³´ê²Œ ë©ë‹ˆë‹¤.
#
# ë˜í•œ, ê²Œì„ì„ ë‹¤ì‹œ ì‹œì‘í•  ë•Œ history ë¦¬ìŠ¤íŠ¸ê°€ ì´ˆê¸°í™”ë˜ì§€ ì•ŠëŠ” ë¬¸ì œë„ í•¨ê»˜ ìˆ˜ì •í–ˆìŠµë‹ˆë‹¤.
```

```python
import random

while True:
    # [ìˆ˜ì • 1] ê²Œì„ì´ ë‹¤ì‹œ ì‹œì‘ë  ë•Œë§ˆë‹¤ ë³€ìˆ˜ë“¤ì„ ì´ˆê¸°í™”í•´ì•¼ í•©ë‹ˆë‹¤.
    # (whileë¬¸ ì•ˆìœ¼ë¡œ ì´ë™)
    history = []
    number = random.randint(1, 50)

    print("\n" + "=" * 40)
    print("ìƒˆë¡œìš´ ê²Œì„ ì‹œì‘! ê¸°íšŒëŠ” 10ë²ˆ")
    print("=" * 40)

    is_success = False

    for i in range(10):  # 0~9 (10ë²ˆ ë°˜ë³µ)
        print(f'\n[{i+1}ë²ˆì§¸ ì‹œë„]')

        # ìˆ«ì ì…ë ¥ ì˜ˆì™¸ ì²˜ë¦¬ (ì„ íƒì‚¬í•­ì´ì§€ë§Œ ê¶Œì¥)
        try:
            guess = int(input('ìˆ«ì ì…ë ¥ (1~50) : '))
        except ValueError:
            print("ìˆ«ìë§Œ ì…ë ¥í•´ì£¼ì„¸ìš”!")
            continue

        history.append(guess)

        if guess == number:
            # iëŠ” 0ë¶€í„° ì‹œì‘í•˜ë¯€ë¡œ íšŸìˆ˜ëŠ” i+1 ì…ë‹ˆë‹¤.
            print(f'ğŸ‰ ì¶•í•˜í•©ë‹ˆë‹¤! {i+1}ë²ˆ ë§Œì— ì •ë‹µ {number}ë¥¼ ë§ì·„ìŠµë‹ˆë‹¤.')

            print("=== í”Œë ˆì´ ê¸°ë¡ ===")
            # ë¦¬ìŠ¤íŠ¸ ì¶œë ¥ë„ íŒŒì´ì¬ìŠ¤ëŸ½ê²Œ ê°„ë‹¨íˆ!
            print(" -> ".join(map(str, history)))

            is_success = True
            print("\nê²Œì„ ë!")
            break  # forë¬¸ íƒˆì¶œ

        elif guess < number:
            print('UP! ë” í° ìˆ«ìì…ë‹ˆë‹¤.')
        else:
            print('DOWN! ë” ì‘ì€ ìˆ«ìì…ë‹ˆë‹¤.')

    # [ê²Œì„ ì¢…ë£Œ í›„ ì²˜ë¦¬]
    # ì‹¤íŒ¨í–ˆì„ ë•Œë§Œ ì •ë‹µ ì•Œë ¤ì£¼ê¸°
    if is_success == False:
        print(f"\nê¸°íšŒë¥¼ ëª¨ë‘ ì‚¬ìš©í–ˆìŠµë‹ˆë‹¤! ë•¡! ì •ë‹µì€ {number}ì˜€ìŠµë‹ˆë‹¤.")

    # [ìˆ˜ì • 2] ì„±ê³µ/ì‹¤íŒ¨ ì—¬ë¶€ì™€ ìƒê´€ì—†ì´ í•­ìƒ ë¬¼ì–´ë´ì•¼ í•˜ë¯€ë¡œ
    # if is_success == False ë¸”ë¡ ë°–ìœ¼ë¡œ ëºìŠµë‹ˆë‹¤.
    retry = input('\në‹¤ì‹œ ì‹œë„í•˜ì‹œê² ìŠµë‹ˆê¹Œ? (y/n) : ')

    if not retry.lower().startswith('y'):
        print('ê²Œì„ì„ ì¢…ë£Œí•©ë‹ˆë‹¤. ì¦ê±°ìš°ì…¨ë‚˜ìš”? ğŸ‘‹')
        break # whileë¬¸ íƒˆì¶œ
```
