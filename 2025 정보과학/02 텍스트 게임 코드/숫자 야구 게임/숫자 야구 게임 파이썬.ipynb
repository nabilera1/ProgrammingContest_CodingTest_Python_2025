````python
import random

# 1. ê²Œì„ ì¤€ë¹„: ì •ë‹µ ìˆ«ì 3ê°œ ìƒì„± (ì¤‘ë³µ ì—†ìŒ)
# "0123456789" ë¬¸ìì—´ ì¤‘ì—ì„œ 3ê°œë¥¼ ë¬´ì‘ìœ„ë¡œ ë½‘ìŠµë‹ˆë‹¤.
# ì˜ˆ: ['1', '5', '9'] í˜•íƒœë¡œ ë¦¬ìŠ¤íŠ¸ê°€ ìƒì„±ë©ë‹ˆë‹¤.
secret_numbers = random.sample("0123456789", 3)

# (í…ŒìŠ¤íŠ¸ìš©) ì •ë‹µì„ ë¯¸ë¦¬ ë³´ê³  ì‹¶ìœ¼ë©´ ì£¼ì„ì„ í•´ì œí•˜ì„¸ìš”
# print(f"ë¹„ë°€ ìˆ«ì: {secret_numbers}")

print("âš¾ ìˆ«ì ì•¼êµ¬ ê²Œì„ì„ ì‹œì‘í•©ë‹ˆë‹¤! âš¾")
print("0ë¶€í„° 9ê¹Œì§€ì˜ ì¤‘ë³µë˜ì§€ ì•ŠëŠ” ìˆ«ì 3ê°œë¥¼ ë§ì¶°ë³´ì„¸ìš”.")
print("-" * 40)

try_count = 0  # ì‹œë„ íšŸìˆ˜

# 2. ê²Œì„ ì‹œì‘ (ë¬´í•œ ë°˜ë³µ)
while True:
    try_count += 1 # ì‹œë„ íšŸìˆ˜ ì¦ê°€

    # ì‚¬ìš©ì ì…ë ¥ ë°›ê¸°
    user_input = input(f"[{try_count}íšŒì°¨] ìˆ«ì 3ê°œë¥¼ ì…ë ¥í•˜ì„¸ìš” > ").strip()

    # --- ì…ë ¥ê°’ ê²€ì¦ (ì˜ëª»ëœ ì…ë ¥ ë°©ì§€) ---
    if len(user_input) != 3:
        print("âŒ ê²½ê³ : 3ìë¦¬ ìˆ«ìë¥¼ ì…ë ¥í•´ì•¼ í•©ë‹ˆë‹¤.")
        continue
    if not user_input.isdigit():
        print("âŒ ê²½ê³ : ìˆ«ìë§Œ ì…ë ¥í•´ì£¼ì„¸ìš”.")
        continue
    if len(set(user_input)) != 3: # setìœ¼ë¡œ ì¤‘ë³µ ì œê±° í›„ ê¸¸ì´ í™•ì¸
        print("âŒ ê²½ê³ : ì¤‘ë³µë˜ëŠ” ìˆ«ìê°€ ìˆìŠµë‹ˆë‹¤.")
        continue

    # --- ìŠ¤íŠ¸ë¼ì´í¬ / ë³¼ íŒì • ë¡œì§ ---
    strikes = 0
    balls = 0

    # enumerateë¥¼ ì‚¬ìš©í•˜ì—¬ ì¸ë±ìŠ¤(i)ì™€ ìˆ«ì(num)ë¥¼ ë™ì‹œì— ê°€ì ¸ì˜´
    for i, num in enumerate(user_input):
        # 1. ìŠ¤íŠ¸ë¼ì´í¬ ì²´í¬: ìˆ«ìì™€ ìœ„ì¹˜ê°€ ëª¨ë‘ ê°™ì„ ë•Œ
        if num == secret_numbers[i]:
            strikes += 1
        # 2. ë³¼ ì²´í¬: ìˆ«ìëŠ” ì •ë‹µì— í¬í•¨ë˜ì–´ ìˆì§€ë§Œ, ìœ„ì¹˜ê°€ ë‹¤ë¥¼ ë•Œ
        elif num in secret_numbers:
            balls += 1

    # --- ê²°ê³¼ ì¶œë ¥ ---
    print(f"ğŸ‘‰ ê²°ê³¼: {strikes} ìŠ¤íŠ¸ë¼ì´í¬, {balls} ë³¼")

    # --- ìŠ¹ë¦¬ ì¡°ê±´ í™•ì¸ ---
    if strikes == 3:
        print("=" * 40)
        print(f"ğŸ‰ ì •ë‹µì…ë‹ˆë‹¤! {try_count}ë²ˆ ë§Œì— ë§ì·„ìŠµë‹ˆë‹¤. ì¶•í•˜í•´ìš”! ğŸ‰")
        print(f"ì •ë‹µ ìˆ«ì: {''.join(secret_numbers)}")
        print("=" * 40)
        break # ê²Œì„ ì¢…ë£Œ (whileë¬¸ íƒˆì¶œ)
````

````python
import random

# 1. ê²Œì„ ì„¤ì •
secret_numbers = random.sample("0123456789", 3)
max_attempts = 10  # ìµœëŒ€ ì‹œë„ íšŸìˆ˜ ì„¤ì •
try_count = 0      # í˜„ì¬ ì‹œë„ íšŸìˆ˜

print(f"âš¾ ìˆ«ì ì•¼êµ¬ ê²Œì„ (HARD MODE) âš¾")
print(f"ê¸°íšŒëŠ” ì´ {max_attempts}ë²ˆì…ë‹ˆë‹¤. ì‹ ì¤‘í•˜ê²Œ ë§ì¶°ë³´ì„¸ìš”!")
print("-" * 40)

# 2. ê²Œì„ ë£¨í”„
while True:
    try_count += 1
    remaining_attempts = max_attempts - try_count # ë‚¨ì€ íšŸìˆ˜ ê³„ì‚°

    # ì‚¬ìš©ì ì…ë ¥
    print(f"\n[ì‹œë„ {try_count}/{max_attempts}] ìˆ«ì 3ê°œë¥¼ ì…ë ¥í•˜ì„¸ìš” (ë‚¨ì€ ê¸°íšŒ: {remaining_attempts}ë²ˆ)")
    user_input = input("ì…ë ¥ > ").strip()

    # --- ì…ë ¥ê°’ ê²€ì¦ ---
    if len(user_input) != 3:
        print("âŒ 3ìë¦¬ ìˆ«ìë¥¼ ì…ë ¥í•´ì•¼ í•©ë‹ˆë‹¤. (íšŸìˆ˜ëŠ” ì°¨ê°ë˜ì§€ ì•ŠìŠµë‹ˆë‹¤)")
        try_count -= 1 # ìœ íš¨í•˜ì§€ ì•Šì€ ì…ë ¥ì´ë¯€ë¡œ íšŸìˆ˜ ë³µêµ¬
        continue
    if not user_input.isdigit():
        print("âŒ ìˆ«ìë§Œ ì…ë ¥í•´ì£¼ì„¸ìš”.")
        try_count -= 1
        continue
    if len(set(user_input)) != 3:
        print("âŒ ì¤‘ë³µë˜ëŠ” ìˆ«ìê°€ ìˆìŠµë‹ˆë‹¤.")
        try_count -= 1
        continue

    # --- íŒì • ë¡œì§ ---
    strikes = 0
    balls = 0

    for i, num in enumerate(user_input):
        if num == secret_numbers[i]:
            strikes += 1
        elif num in secret_numbers:
            balls += 1

    # --- ê²°ê³¼ ì¶œë ¥ ---
    print(f"ğŸ‘‰ ê²°ê³¼: {strikes} ìŠ¤íŠ¸ë¼ì´í¬, {balls} ë³¼")

    # --- [ìŠ¹ë¦¬] 3 ìŠ¤íŠ¸ë¼ì´í¬ë©´ ê²Œì„ ì¢…ë£Œ ---
    if strikes == 3:
        print("\n" + "=" * 40)
        print(f"ğŸ‰ ì¶•í•˜í•©ë‹ˆë‹¤! {try_count}ë²ˆ ë§Œì— í™ˆëŸ°ì„ ì³¤ìŠµë‹ˆë‹¤! ğŸ‰")
        print("=" * 40)
        break

    # --- [íŒ¨ë°°] ê¸°íšŒë¥¼ ëª¨ë‘ ì†Œì§„í–ˆëŠ”ì§€ í™•ì¸ ---
    if try_count >= max_attempts:
        print("\n" + "=" * 40)
        print("ğŸ’€ GAME OVER! ëª¨ë“  ê¸°íšŒë¥¼ ì†Œì§„í–ˆìŠµë‹ˆë‹¤.")
        print(f"ì •ë‹µì€ [{' '.join(secret_numbers)}] ì˜€ìŠµë‹ˆë‹¤.")
        print("=" * 40)
        break
````
