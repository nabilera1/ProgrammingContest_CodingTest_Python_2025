# 반복문


````python
코딩 기초 실력을 튼튼하게 다질 수 있는 쉬운 연습 문제 3개를 준비했습니다.
조건문(if), 반복문(for), 리스트를 활용하는 핵심 예제들입니다.

1. 1부터 10까지 짝수만 더하기
숫자 1부터 10 사이의 숫자 중에서 짝수(2, 4, 6, 8, 10)만 골라서 합계를 구해보세요.

힌트: 나머지를 구하는 % 연산자를 사용하세요. (i % 2 == 0이면 짝수)

정답 코드:


total = 0

for i in range(1, 11):  # 1부터 10까지 반복
    if i % 2 == 0:      # 짝수라면
        total += i      # 합계에 더하기

print(f"짝수의 합: {total}")
# 결과: 30


2. 거꾸로 말해요 (문자열 뒤집기)
입력받은 단어를 거꾸로 출력하는 코드를 작성하세요. (예: "apple" → "elppa")

힌트: 파이썬의 슬라이싱 [::-1] 기능을 쓰면 아주 쉽습니다.

정답 코드:


word = "Python"

# 슬라이싱 [시작:끝:간격]에서 간격을 -1로 주면 뒤집힙니다.
reversed_word = word[::-1]

print(f"원본: {word}")
print(f"거꾸로: {reversed_word}")
# 결과: nohtyP


3. 가장 큰 숫자 찾기 (max 함수 없이)
리스트에 있는 숫자들 중에서 가장 큰 숫자를 찾아보세요.
단, 파이썬의 max() 기능을 쓰지 않고 for문으로 직접 만들어보세요.

힌트: 첫 번째 숫자를 '현재 1등'이라고 가정하고,
다른 숫자들과 비교하며 더 큰 숫자가 나오면 '1등'을 바꾸면 됩니다.

정답 코드:


numbers = [20, 55, 10, 99, 5]

max_num = numbers[0]  # 일단 첫 번째 숫자를 가장 크다고 가정

for n in numbers:
    if n > max_num:   # 현재 잡고 있는 숫자보다 더 큰 숫자를 만나면
        max_num = n   # 그 숫자로 교체

print(f"가장 큰 수: {max_num}")
# 결과: 99


def max1(num_list):
    # 1. 첫 번째 숫자를 기준(챔피언)으로 잡음
    max_num = num_list[0]

    # 2. 리스트를 돌면서 비교
    for n in num_list:
        if n > max_num:   # 더 큰 숫자를 만나면
            max_num = n   # 챔피언 교체

    # 3. 최종적으로 남은 가장 큰 수를 돌려줌
    return max_num

# --- 함수 사용해보기 ---
numbers = [20, 55, 10, 99, 5]

# 함수를 호출해서 결과를 result 변수에 받음
result = max1(numbers)

print(f"가장 큰 수: {result}")




💡 도전 과제: 3번 문제(가장 큰 숫자 찾기)를 응용해서
"가장 작은 숫자 찾기"도 만들어 보실 수 있나요? (부등호 방향만 바꾸면 됩니다!)
````

````python
파이썬의 for문과 문자열 연산(곱하기 *)을 활용하면 별 찍기 문제를 아주 간결하게 해결할 수 있습니다.

난이도별로 5단계를 준비했습니다. 눈으로 먼저 로직을 생각해보고 코드를 확인해 보세요!

[Lv.1] 왼쪽으로 치우친 삼각형
가장 기초적인 문제입니다. 줄(i)이 늘어날수록 별도 하나씩 늘어납니다.

목표 출력:

*
**
***
****
*****
정답 코드:


N = 5
for i in range(1, N + 1):
    print("*" * i)
해설: range(1, 6)은 1부터 5까지 반복합니다. "*" 문자에 숫자 i를 곱하면 별이 i개만큼 찍힙니다.

[Lv.2] 거꾸로 역삼각형
이번에는 별이 5개부터 시작해서 1개로 줄어듭니다. range를 거꾸로 돌리는 것이 핵심입니다.

목표 출력:

*****
****
***
**
*
정답 코드:



N = 5
for i in range(N, 0, -1):  # 5부터 1까지 -1씩 줄어듦
    print("*" * i)
해설: range(시작, 끝, 간격)에서 간격을 -1로 주면 숫자가 줄어듭니다.

[Lv.3] 오른쪽 정렬 삼각형 (공백의 마법)
오른쪽에 붙으려면 "공백"을 먼저 찍고, 그 뒤에 "별"을 찍어야 합니다.

목표 출력:

    *
   **
  ***
 ****
*****


정답 코드:



N = 5
for i in range(1, N + 1):
    spaces = " " * (N - i)  # 공백 개수: 4, 3, 2, 1, 0
    stars = "*" * i         # 별 개수: 1, 2, 3, 4, 5
    print(spaces + stars)
해설: 전체 5칸 중에서 별이 1개면 공백은 4개, 별이 2개면 공백은 3개... 즉, 공백 = 전체(N) - 별(i) 공식이 성립합니다.

[Lv.4] 정삼각형 (피라미드)
좌우 대칭을 맞추는 문제입니다. 별의 개수가 1, 3, 5, 7, 9개로 **홀수(odd)**로 늘어나는 규칙을 찾아보세요.

목표 출력:

    *
   ***
  *****
 *******
*********
정답 코드:

Python

N = 5
for i in range(1, N + 1):
    spaces = " " * (N - i)
    stars = "*" * (2 * i - 1)  # 홀수 공식: 2*i - 1
    print(spaces + stars)
해설:

1번째 줄: 별 1개 (2*1 - 1)

2번째 줄: 별 3개 (2*2 - 1)

3번째 줄: 별 5개 (2*3 - 1)

이 규칙(2*i - 1)을 찾는 것이 알고리즘의 핵심입니다.

[Lv.5] 다이아몬드 (중급 보스)
피라미드를 정방향으로 한 번, 역방향으로 한 번 찍으면 됩니다. (가운데 가장 긴 줄이 겹치지 않게 조절하는 디테일이 필요합니다!)

목표 출력:

    *
   ***
  *****
   ***
    *
정답 코드:

Python

N = 3  # 중심까지의 높이

# 1. 윗부분 (피라미드)
for i in range(1, N + 1):
    print(" " * (N - i) + "*" * (2 * i - 1))

# 2. 아랫부분 (역피라미드)
for i in range(N - 1, 0, -1):  # N-1부터 시작해야 겹치지 않음
    print(" " * (N - i) + "*" * (2 * i - 1))
해설:

윗부분: 1부터 N(3)까지 별이 늘어납니다.

아랫부분: N-1(2)부터 1까지 별이 줄어듭니다.

이 두 개의 for문을 순서대로 배치하면 다이아몬드가 완성됩니다.

이 문제들을 혼자 힘으로 풀 수 있다면, 반복문과 문자열 처리 기초는 마스터하신 겁니다! 🥳
````
