```python
import random
import time

class WordQuizGame:
    """
    random ëª¨ë“ˆê³¼ ë¬¸ìì—´ ì²˜ë¦¬ ë©”ì„œë“œë¥¼ í™œìš©í•œ ì˜ë‹¨ì–´ ë§ì¶”ê¸° ê²Œì„ í´ë˜ìŠ¤
    """

    def __init__(self):
        # í€´ì¦ˆ ë°ì´í„° (ì¹´í…Œê³ ë¦¬: [ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸])
        self.word_db = {
            "ê³¼ì¼": ["APPLE", "BANANA", "GRAPE", "ORANGE", "PEACH"],
            "ë™ë¬¼": ["TIGER", "LION", "RABBIT", "ELEPHANT", "PANDA"],
            "í”„ë¡œê·¸ë˜ë°": ["PYTHON", "JAVA", "ALGORITHM", "VARIABLE", "STRING"]
        }
        self.max_attempts = 7  # ìµœëŒ€ ì‹œë„ íšŸìˆ˜

    def select_word(self):
        """ì¹´í…Œê³ ë¦¬ì™€ ë‹¨ì–´ë¥¼ ë¬´ì‘ìœ„ë¡œ ì„ íƒ"""
        category = random.choice(list(self.word_db.keys()))
        word = random.choice(self.word_db[category])
        return category, word

    def get_masked_word(self, word, guessed_letters):
        """
        ë§ì¶˜ ê¸€ìëŠ” ë³´ì—¬ì£¼ê³ , ëª» ë§ì¶˜ ê¸€ìëŠ” ì–¸ë”ë°”(_)ë¡œ í‘œì‹œ
        (ë¬¸ìì—´ join ë©”ì„œë“œì™€ ë¦¬ìŠ¤íŠ¸ ì»´í”„ë¦¬í—¨ì…˜ í™œìš©)
        """
        display_list = [char if char in guessed_letters else "_" for char in word]
        return " ".join(display_list)  # ë¦¬ìŠ¤íŠ¸ë¥¼ ë¬¸ìì—´ë¡œ ê²°í•© (ì˜ˆ: "A _ P L _")

    def print_header(self, title):
        """í™”ë©´ ìƒë‹¨ ê¾¸ë¯¸ê¸° (center ë©”ì„œë“œ í™œìš©)"""
        print("=" * 40)
        print(title.center(40))  # 40ì¹¸ í¬ê¸°ì—ì„œ ê°€ìš´ë° ì •ë ¬
        print("=" * 40)

    def play(self):
        """ê²Œì„ ì‹¤í–‰ ë©”ì¸ ë¡œì§"""
        category, secret_word = self.select_word()
        guessed_letters = set()  # ì¤‘ë³µ ë°©ì§€ë¥¼ ìœ„í•´ set ì‚¬ìš©
        attempts = 0

        self.print_header("ğŸ¯ ì˜ë‹¨ì–´ í€´ì¦ˆ ê²Œì„ ì‹œì‘! ğŸ¯")
        print(f"ì¹´í…Œê³ ë¦¬: [{category}]")
        print(f"ì´ {len(secret_word)}ê¸€ìì˜ ë‹¨ì–´ì…ë‹ˆë‹¤.\n")

        while attempts < self.max_attempts:
            # 1. í˜„ì¬ ìƒíƒœ ì¶œë ¥
            current_display = self.get_masked_word(secret_word, guessed_letters)
            print(f"\në‹¨ì–´: {current_display}")
            print(f"ë‚¨ì€ ê¸°íšŒ: {self.max_attempts - attempts}ë²ˆ")

            # 2. ì‚¬ìš©ì ì…ë ¥ ë°›ê¸° ë° ì „ì²˜ë¦¬ (strip, upper í™œìš©)
            user_input = input("ì•ŒíŒŒë²³ í•˜ë‚˜ë¥¼ ì…ë ¥í•˜ì„¸ìš” > ").strip().upper()

            # 3. ì…ë ¥ê°’ ê²€ì¦ (isalpha, len í™œìš©)
            if len(user_input) != 1:
                print("âŒ ê²½ê³ : í•œ ê¸€ìë§Œ ì…ë ¥í•´ì£¼ì„¸ìš”.")
                continue
            if not user_input.isalpha():
                print("âŒ ê²½ê³ : ì•ŒíŒŒë²³(ì˜ì–´)ë§Œ ì…ë ¥ ê°€ëŠ¥í•©ë‹ˆë‹¤.")
                continue
            if user_input in guessed_letters:
                print(f"âš ï¸ ì´ë¯¸ ì…ë ¥í•œ ê¸€ìì…ë‹ˆë‹¤: {user_input}")
                continue

            # 4. ê²Œì„ ë¡œì§ ì²˜ë¦¬
            guessed_letters.add(user_input)

            # ì‹œê°„ ì§€ì—° íš¨ê³¼ (ê²Œì„ì˜ ê¸´ì¥ê° ì¡°ì„±)
            print("í™•ì¸ ì¤‘...", end="")
            time.sleep(0.5)

            if user_input in secret_word:
                print(f"\râœ… ì •ë‹µì…ë‹ˆë‹¤! '{user_input}'ê°€ í¬í•¨ë˜ì–´ ìˆìŠµë‹ˆë‹¤.")

                # ëª¨ë“  ê¸€ìë¥¼ ë‹¤ ë§ì·„ëŠ”ì§€ í™•ì¸ (ê³µë°± ì—†ì´ ë¬¸ìì—´ í•©ì³ì„œ ë¹„êµ)
                # replace(" ", "")ë¥¼ ì¨ì„œ "A P P L E" -> "APPLE"ë¡œ ë³€í™˜
                if self.get_masked_word(secret_word, guessed_letters).replace(" ", "") == secret_word:
                    self.print_header("ğŸ‰ ì¶•í•˜í•©ë‹ˆë‹¤! ì •ë‹µì„ ë§ì·„ìŠµë‹ˆë‹¤! ğŸ‰")
                    print(f"ì •ë‹µ ë‹¨ì–´: {secret_word}")
                    break
            else:
                attempts += 1
                print(f"\râŒ ë•¡! '{user_input}'ëŠ” ì—†ìŠµë‹ˆë‹¤.")

        # ë°˜ë³µë¬¸ì´ break ì—†ì´ ëë‚¬ë‹¤ë©´ (ê¸°íšŒ ì†Œì§„)
        if attempts == self.max_attempts:
            self.print_header("ğŸ’€ ê²Œì„ ì˜¤ë²„ ğŸ’€")
            print(f"ì •ë‹µì€ [{secret_word}] ì˜€ìŠµë‹ˆë‹¤.")


if __name__ == "__main__":
    # ê²Œì„ ë°˜ë³µ ì‹¤í–‰ ë¡œì§
    while True:
        game = WordQuizGame()
        game.play()

        # ì¬ì‹œì‘ ì—¬ë¶€ í™•ì¸ (lower, startswith í™œìš©)
        retry = input("\në‹¤ì‹œ í•˜ì‹œê² ìŠµë‹ˆê¹Œ? (y/n): ").strip().lower()
        if not retry.startswith('y'):
            print("\nê²Œì„ì„ ì¢…ë£Œí•©ë‹ˆë‹¤. ì¦ê±°ìš´ í•˜ë£¨ ë˜ì„¸ìš”!")
            break
        print("\n" * 2)  # í™”ë©´ ë¶„ë¦¬
```
