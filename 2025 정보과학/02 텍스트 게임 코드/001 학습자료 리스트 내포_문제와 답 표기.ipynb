### Q2. ì¡°ê±´ í•„í„°ë§ (if)

### Q3. ë¬¸ìì—´ ê¸¸ì´ í•„í„°ë§
ì•„ë˜ ë¦¬ìŠ¤íŠ¸ì—ì„œ **ê¸€ì ìˆ˜ê°€ 4 ì´ìƒ**ì¸ ë‹¨ì–´ë§Œ ì¶”ì¶œí•˜ì„¸ìš”.

### Q6. ë³µí•© ë¡œì§ (ê³„ì‚° + í•„í„°ë§)

### Q8. ë”•ì…”ë„ˆë¦¬ ë‚´í¬ (Key-Value êµí™˜)
