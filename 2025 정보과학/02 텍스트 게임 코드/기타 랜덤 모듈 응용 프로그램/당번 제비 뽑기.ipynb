당번 N명 뽑기 (제비뽑기 로직)

역할 구분 없이 "오늘 남아서 청소할 사람 3명!"을 뽑을 때 사용하는 sample 활용 코드입니다.

```python
import random

students = ["철수", "영희", "민수", "지수", "현우", "민지", "나영"]

# 당번 인원 설정
count = 3

print("🥁 두근두근... 오늘의 청소 당번 추첨 🥁")

# sample(리스트, 개수): 중복 없이 n명을 뽑음
lucky_members = random.sample(students, count)

# 결과 출력 (보기 좋게 쉼표로 연결)
print(f"당첨자: {', '.join(lucky_members)}")
print(f"축하합니다(?) 오늘 교실을 깨끗이 부탁해요!")
```

선생님을 위한 코칭 팁

zip() 함수 설명: "역할 리스트와 학생 리스트를 지퍼 잠그듯이 서로 짝지어 주는 함수"라고 설명하면 학생들이 직관적으로 이해합니다.

shuffle() vs sample():

Shuffle: 전체 순서를 섞어서 모두에게 순위/역할을 줄 때 (예: 발표 순서, 자리 배치).

Sample: 전체 중 일부만 뽑을 때 (예: 청소 당번, 경품 추첨).

응용 과제: "매일 당번을 뽑으면 걸린 사람이 또 걸릴 수 있죠? 이를 방지하려면 어떻게 해야 할까요?"라는 질문을 던져보세요. (답: 뽑힌 사람은 리스트에서 remove 하고 파일로 저장한다 등)

```python
선생님, 말씀하신 "한 번 당번을 한 학생은 명단에서 제외하고, 전원이 다 하면 다시 리셋되는" 공정한 로테이션 시스템 코드입니다.

이 로직은 **상태 관리(State Management)**의 기초가 되는 개념이라, 프로그래밍적 사고를 기르는 데 아주 좋습니다.

🧹 공정한 청소 당번 로테이션 (중복 방지 시스템)
핵심 로직:

**후보 명단(candidates)**을 따로 관리합니다.

당번으로 뽑힌 학생은 후보 명단에서 **제거(remove)**합니다.

후보 명단이 부족하면, 원본 명단에서 다시 채워 넣습니다(리셋).
```

```python
import random
import time

class CleaningRotation:
    def __init__(self, students):
        self.all_students = students  # 원본 명단 (변하지 않음)
        self.candidates = list(students) # 현재 후보 명단 (계속 줄어듦)
# 그냥 candidates = students라고 하면 얕은 복사(Shallow Copy)가 되어, 후보를 지울 때 원본도 같이 지워지는 참사가 일어납니다. 반드시 list()를 씌워 새로운 리스트로 복사해야 함을 강조해 주세요.
    def pick_cleaners(self, count):
        print(f"\n[현재 남은 후보]: {len(self.candidates)}명 {self.candidates}")

        # 1. 후보가 부족하면 리셋 (로테이션 종료)
        if len(self.candidates) < count:
            print("🔄 모든 학생이 청소를 완료했습니다! 명단을 초기화합니다.")
            self.candidates = list(self.all_students)
            print(f"(명단 복구 완료: {len(self.candidates)}명)")

        # 2. 후보 중에서 무작위 추첨
        picked = random.sample(self.candidates, count)

        # 3. 뽑힌 학생을 후보 명단에서 제거 (핵심 로직)
        for student in picked:
            self.candidates.remove(student)

        return picked

# --- 실행 시뮬레이션 ---

# 전체 학생 명단
student_list = ["철수", "영희", "민수", "지수", "호영", "민지", "다혜"]
manager = CleaningRotation(student_list)

# 매일 2명씩 5일 동안 뽑는 상황 가정
day = 1
while True:
    input(f"\n📅 {day}일차 당번을 뽑으려면 Enter를 누르세요...")

    # 당번 2명 선정
    todays_cleaners = manager.pick_cleaners(2)

    print(f"🎉 오늘의 당번: {todays_cleaners}")
    print("-" * 40)

    day += 1

    # (테스트를 위해 5일차까지만 반복)
    if day > 5:
        print("\n✅ 시뮬레이션을 종료합니다.")
        break
```

선생님을 위한 포인트
self.candidates = list(students):

이 부분이 중요합니다. 그냥 candidates = students라고 하면 **얕은 복사(Shallow Copy)**가 되어, 후보를 지울 때 원본도 같이 지워지는 참사가 일어납니다. 반드시 list()를 씌워 새로운 리스트로 복사해야 함을 강조해 주세요.

remove() 메서드:

리스트에서 특정 값을 찾아 지우는 메서드입니다. 이 코드를 통해 "데이터를 지워나가는 과정"을 시각적으로 보여줄 수 있습니다.

이 코드는 실제로 교실 컴퓨터에 켜두고 매일 아침마다 실행해서 당번을 정하는 용도로도 쓸 수 있습니다! (물론 프로그램 끄면 데이터가 날아가니, 파일 저장 기능까지 배우면 완벽한 프로그램이 됩니다.)
