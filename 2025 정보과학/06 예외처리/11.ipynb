```python
# try_else.py
try:
    age=int(input('나이를 입력하세요: '))
except:
    print('입력이 정확하지 않습니다.')
else:
    if age <= 18:
        print('미성년자는 출입금지입니다.')
    else:
        print('환영합니다.')
finally:
    print('무조건 실행되는 부분')
```

```python
class MyError(Exception):
    def __str__(self):
        return 'MyError: 허용되지 않는 별명'

def say_nick(nick):
    if nick=='바보':
        raise MyError
    else:
        print(nick)

try:
    say_nick("천사")
    say_nick("바보")
except MyError as e:
    print(e) # e.__str__
```

```python
scores = [85, 92, 78, 95, 88]
attendance_check = [True, True, False, True, True]
names = ["A 학생", "B 학생", "C 학생", "D 학생", "E 학생"]

# 모든 학생이 80점 이상인가요?  체크하는 코드를 한줄로 작성하시오.
all_passed = all(s >= 80 for s in scores)
print(all_passed)
```

```python
[s >= 80 for s in scores]
```

```python
# 모든 학생이 출석했나요?
all_attended = all(attendance_check)
print(all_attended)
```

```python
# 100점 받은 학생이 한 명이라도 있나요
any_perfect = any(s==100 for s in scores)
print(any_perfect)
```

```python
scores = [85, 92, 78, 95, 88]
attendance_check = [True, True, False, True, True]
names = ["A 학생", "B 학생", "C 학생", "D 학생", "E 학생"]
```
