```python
students = ["김철수", "이영희", "박민수", "최유진"]

for student in students:
    try:
        with open(f"{student}_성적.txt", 'r') as f:
            score = f.read()
            print(f"{student}의 성적: {score}")
    except FileNotFoundError:
        print(f"{student}의 성적 파일이 없습니다. 건너뜁니다.")
        continue  # 다음 학생으로 넘어감
```

```python
class MyError(Exception):
    def __str__(self):
        return 'MyError: 허용되지 않는 별명'

def say_nick(nick):
    if nick == "바보":
        raise MyError
    else:
        print(nick)

# say_nick("천사")
try:
    say_nick("바보")
except MyError as e:
    # print("NO~~~")
    print(e) # e.__str__
```

```python
%%sql
def binary_search(arr, target):
    lo, hi = 0, len(arr) - 1
    while lo <= hi:
        mid = (lo + hi) // 2
        if arr[mid] == target:
            return mid
        if arr[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1
    return -1

arr = [1, 3, 4, 7, 9, 11, 15]
print(binary_search(arr, 7))   # 3
print(binary_search(arr, 8))   # -1
```
