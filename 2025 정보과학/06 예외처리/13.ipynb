```
students = ["김철수", "이영희", "박민수", "최유진"]

for student in students:
    try:
        with open(f"{student}_성적.txt", 'r') as f:
            score = f.read()
            print(f"{student}의 성적: {score}")
    except FileNotFoundError:
        print(f"{student}의 성적 파일이 없습니다. 건너뜁니다.")
        continue  # 다음 학생으로 넘어감
```

```
class Bird:
    def fly(self):
        raise NotImplementedError

class Eagle(Bird):
    def fly(self):
        print('파닥파닥 펄럭펄럭')

e = Eagle()
e.fly()
```

```
class MyError(Exception):
    def __str__(self):
        return 'MyError: 허용되지 않는 별명'

def say_nick(nick):
    if nick == "바보":
        raise MyError
    else:
        print(nick)

# say_nick("천사")
try:
    say_nick("바보")
except MyError as e:
    # print("NO~~~")
    print(e) # e.__str__
```
