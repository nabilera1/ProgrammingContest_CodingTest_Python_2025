```python
# error_make.py
class MyError(Exception):
    pass
def say_nick(nick):
    if nick == '바보':
        raise MyError()
    print(nick)
say_nick("천사")
say_nick("바보")
```

```python
try:
    say_nick("천사")
    say_nick("바보")
except MyError:
    print("허용되지 않는 별명입니다.")

```

```python
class MyError(Exception):
    def __str__(self):
        return "허용되지 않는 별명 (str)"
```

```python
try:
    say_nick("천사")
    say_nick("바보")
except MyError as e:
    print(e)
```

```python
print(" 사탕 나누기 기계 가동!")

candy = 10
friends = 2  #  친구가 0명이면? (이 숫자를 2나 5로 바꿔보기기)

try:
    # [시도] 사탕 나누기 시도!
    result = candy / friends

except ZeroDivisionError:
    # [실패] 0으로 나누려고 할 때
    print("앗! 0명한테는 사탕을 나눠줄 수가 없어요. ")

else:
    # [성공] 에러 없이 잘 나누어졌을 때만 실행
    print(f"와! 한 명당 사탕을 {result}개씩 먹으면 돼요! ")

finally:
    print("기계 전원을 끕니다.")
```

```python
print(" 자판기 앞에 섰습니다.")

menu = ["콜라", "사이다", "주스"]
button = 1  #  메뉴는 0,1,2번 뿐인데 5번을 눌렀어요!

try:
    # [시도] 버튼에 맞는 음료수 꺼내기
    drink = menu[button]

except IndexError:
    # [실패] 없는 번호를 눌렀을 때
    print("삐-익! 그런 음료수는 없어요. 다시 눌러주세요. ")

else:
    # [성공] 음료수가 잘 나왔을 때만 실행
    print(f"덜컹! {drink}가 나왔습니다. 꿀꺽꿀꺽 마셔요!")

finally:
     print("자판기를 이용해주셔서 감사합니다. ")
```
