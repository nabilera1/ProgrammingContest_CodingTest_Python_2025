```python
scores = [85, 92, 78, 95, 88]
attendance_check = [True, True, False, True, True] # bool
names = ["A 학생", "B 학생", "C 학생", "D 학생", "E 학생"]
```

```python
# any, 100점 받은 학생이 한 명이라도 있는지 체크
any_perfect = any(s == 100 for s in scores)
print(any_perfect)
```

```python
# 결석한 학생이 한 명이라도 있는지 체크
any_absent = any(not a for a in attendance_check)
print(any_absent)
```

```python
#enumerate

for i, name in enumerate(names, start=1):
    score = scores[i-1]
    print(f"{i}. {name}: {score}")
```

```python
def binary_search(arr, target):
    lo, hi = 0, len(arr) - 1
    while lo <= hi:
        mid = (lo + hi) // 2
        if arr[mid] == target:
            return mid
        if arr[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1
    return -1

arr = [1, 3, 4, 7, 9, 11, 15]
print(binary_search(arr, 7))   # 3
print(binary_search(arr, 8))   # -1
```
