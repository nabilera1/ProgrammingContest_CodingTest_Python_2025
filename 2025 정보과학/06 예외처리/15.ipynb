```python
try:
    age=int(input('나이를 입력하세요: '))
except:
    print('입력이 정확하지 않습니다.')
else:
    if age <= 18:
        print('미성년자는 출입금지입니다.')
    else:
        print('환영합니다.')
finally:
    print('반드시 실행되어야하는 코드는 여기에 작성')
```

```python
students = ["김철수", "이영희", "박민수", "최유진"]

for student in students:
    try:
        with open(f"{student}_성적.txt", 'r') as f:
            score = f.read()
            print(f"{student}의 성적: {score}")
    except FileNotFoundError:
        print(f"{student}의 성적 파일이 없습니다. 건너뜁니다.")
        #continue  # 다음 학생으로 넘어감
```

```python
class MyError(Exception):
    pass

def say_nick(nick):
    if nick == "바보":
        raise MyError
    print(nick)

say_nick('천사')
```

```python
try:
    say_nick('바보')
except MyError as e:
    print('예외 처리됨')
```

```python
class MyError(Exception):
    def __str__(self):
        return "MyError: 바보 아니야~"

try:
    say_nick('천사')
    say_nick('바보')
except MyError as e:
    print(e) # e.__str__
```

```python
print(" 사탕 나누기 기계 가동!")

candy = 10
friends = 2  #  친구가 0명이면? (이 숫자를 2나 5로 바꿔보기기)

try:
    result = candy / friends
except ZeroDivisionError:
    print("앗! 0명한테는 사탕을 나눠줄 수가 없어요. ")
else:
    print(f"와! 한 명당 사탕을 {result}개씩 먹으면 돼요! ")
finally:
    print("기계 전원을 끕니다.")
```

```python
print(" 자판기 앞에 섰습니다.")

menu = ["콜라", "사이다", "주스"]
button = 4  #  메뉴는 0,1,2번 뿐인데 5번을 눌렀어요!

try:
    drink = menu[button]
except IndexError:
    print("삐-익! 그런 음료수는 없어요. 다시 눌러주세요. ")
else:
    print(f"덜컹! {drink}가 나왔습니다. 꿀꺽꿀꺽 마셔요!")
finally:
     print("자판기를 이용해주셔서 감사합니다. ")
```
