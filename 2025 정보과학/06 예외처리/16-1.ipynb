```python
scores = [85, 92, 78, 95, 88]
attendance_check = [True, True, False, True, True]
names = ["A 학생", "B 학생", "C 학생", "D 학생", "E 학생"]

# 모든 학생이 80점 이상인지 체크
all_passed = all(s>80 for s in scores)
print(all_passed)

print([s>80 for s in scores])
```

```python
# 모든 학생이 출석했는지 체크
all_attended = all(attendance_check)
print(all_attended)
```

```python
# any(x)는 반복 가능한 데이터 x를 입력으로 받아 x의 요소 중 하나라도 참이 있으면 True를 반환하고 x가 모두 거짓일 때만 False를 반환한다. 즉, all(x)의 반대로 작동한다.
```

```python
# 100점 받은 학생이 한 명이라도 있는가?
any_perfect = any(s==100 for s in scores)
print(any_perfect)
```

```python
# 결석한 학생이 한 명이라도 있나요?
scores = [85, 92, 78, 95, 88]
attendance_check = [True, True, False, True, True]
names = ["A 학생", "B 학생", "C 학생", "D 학생", "E 학생"]
any_absent = any(s==False for s in attendance_check)
print(any_absent)
```

```python
any_absent1 = any(not a for a in attendance_check)
print(any_absent1)
```

```python
def binary_search(arr, target):
    lo, hi = 0, len(arr) - 1
    while lo <= hi:
        mid = (lo + hi) // 2
        if arr[mid] == target:
            return mid
        if arr[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1
    return -1

arr = [1, 3, 4, 7, 9, 11, 15]
print(binary_search(arr, 7))   # 3
print(binary_search(arr, 8))   # -1
```
