## 시험 통계 구하기 (딕셔너리 안의 리스트)

## 단어 등장 횟수 Top 3 (리스트 + 딕셔너리 + 정렬)

```python
freq = {}

for w in ["dog", "cat", "dog"]:
    freq[w] = freq.get(w, 0) + 1   # 있으면 가져오고, 없으면 0부터 시작

print(freq)
```

```python
text = "dog cat dog dog cat bird dog cat bird bird bird"
words = text.split()

freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1

# 개수 내림차순, 동률이면 처음 등장 순서를 유지하고 싶다면 Counter 사용 권장
top3 = sorted(freq.items(), key=lambda x: x[1], reverse=True)[:3]

for w, c in top3:
    print(w, c)
```

```python
from collections import Counter

text = "dog cat dog dog cat bird"
words = text.split()

counter = Counter(words)
print(counter)
print(type(counter))
print(counter.most_common(2))
```

```python
from collections import Counter

text = "dog cat dog dog cat bird dog cat bird bird bird"
counter = Counter(text.split())

for w, c in counter.most_common(3):
    print(w, c)
```

```python
import random

# 조건문·반복문·리스트·문자열 학습하기
HANGMAN = [
'''
  +---+
  |   |
      |
      |
      |
      |
=========''',
'''
  +---+
  |   |
  O   |
      |
      |
      |
=========''',
'''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========''',
'''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''',
'''
  +---+
  |   |
  O   |
 /|\\  |
      |
      |
=========''',
'''
  +---+
  |   |
  O   |
 /|\\  |
 /    |
      |
=========''',
r'''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
========='''
]


# 단어 목록
words = ["apple", "banana", "grape", "orange", "melon"]
secret = random.choice(words) # 랜덤 선택
# print(secret)
missed = ""
correct = ""
print("==== H A N G M A N ====")

while True:
  print(HANGMAN[len(missed)])
  display =""
  # for ch in secret:
  #   if ch in correct:
  #     display += ch
  #   else:
  #     display += "_"
  for ch in secret:
    display += ch if ch in correct else "_"

  print('Word : ', " ".join(display)) # _ _ _ _ _
  print('Missed : ', " ".join(missed))

  if display == secret:
    print("You Win!")
    break
  if len(missed) == len(HANGMAN) - 1:
    print('Game over! The answer is', secret)
    break

  guess = input('Guess a letter: ')
  if guess in secret:
    correct += guess
  else:
    missed += guess

```

# 6 팩토리얼 함수

```python
def fact(n):
  if n<=1: return 1
  return n * fact(n-1)

def fact_iter(n):
  r = 1
  for i in range(1, n+1):
    r *= i
  return r

memo = {0:1, 1: 1}
def fact_memo(n):
  if n not in memo:
    memo[n] = n*fact_memo(n-1)
  return memo[n]
```
