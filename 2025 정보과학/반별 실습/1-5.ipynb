```python
import random

# 조건문·반복문·리스트·문자열 학습하기
HANGMAN = [
'''
  +---+
  |   |
      |
      |
      |
      |
=========''',
'''
  +---+
  |   |
  O   |
      |
      |
      |
=========''',
'''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========''',
'''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''',
'''
  +---+
  |   |
  O   |
 /|\\  |
      |
      |
=========''',
'''
  +---+
  |   |
  O   |
 /|\\  |
 /    |
      |
=========''',
r'''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
========='''
]


# 단어 목록
words = ["apple", "banana", "grape", "orange", "melon"]
secret = random.choice(words) # 랜덤 선택
# print(secret)
missed = ""
correct = ""
print("==== H A N G M A N ====")
while True:
    print(HANGMAN[len(missed)])
    display = ""
    for ch in secret:
        display += ch if ch in correct else "_"

    print('Word:', " ".join(display)) #_ _ _ _ _
    print('Missed:', " ".join(missed))

    if display == secret:
        print('Correct!')
        break
    if len(missed) == len(HANGMAN) -1:
        print('Game Over! The secret word was:', secret)
        break

    guess = input('Guess a letter: ')
    if guess in secret:
        correct += guess
    else:
        missed += guess
```
